# Divisibility by 5

We are finding for a rule that helps us determine if a number is `fully divisible by 5`
- We will shorten that phrase to `divisible by 5` instead
- Because `technically every number is divisble by 5` even if the `quotient is irrational or transcendental`, so it has `no meaningful meaning`
- Instead, it is remapped to mean `fully divisible` instead


# Decomposition of a Number

Any number can be represented in its `base-10` form like so
- Where the variables range from `0` to `9`

$$
\text{Number} \space = \space \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n places}}
$$

When each place is expanded out, we get 

$$
\underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n places}} \space = 
\space \underline{k \cdot 10^{n}} + 
\underline{m \cdot 10^{n-1}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

Thus

$$
\text{Number} \space =
\space \underline{k \cdot 10^{n}} + 
\underline{m \cdot 10^{n-1}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

In order for this to be `divisible by 5`, we must be able to factor out `9` from this `expression`

Notice that we can factor this number from `2 different parts`
- One is the `place values` of the number

$$
\underline{k \textcolor{lightgray}{\cdot 10^{n}}} + 
\underline{m \textcolor{lightgray}{\cdot 10^{n-1}}} + 
\cdots +
\underline{c \textcolor{lightgray}{\cdot 10^{2}}} + 
\underline{b \textcolor{lightgray}{\cdot 10^{1}}} + 
\underline{a \textcolor{lightgray}{\cdot 10^{0}}}
$$

- The other is the `powers of 10` that represents the `grouping in each place` 

$$
\underline{\textcolor{lightgray}{k \space \cdot} \space 10^{n}} + 
\underline{\textcolor{lightgray}{m \space \cdot} \space 10^{n-1}} + 
\cdots + 
\underline{\textcolor{lightgray}{c \space \cdot} \space 10^{2}} + 
\underline{\textcolor{lightgray}{b \space \cdot} \space 10^{1}} + 
\underline{\textcolor{lightgray}{a \space \cdot} \space 10^{0}}
$$

We can't do much for the `place values` since they are all just variables, 

> But we can look into the `places` itself!

# Factoring Powers of 10

Lets start from $10^0$
- Which equals `1`

$$
10^{0} = 1
$$

We cannot `partition` nor `factor` this that so lets move on to $10^1$

## Intentional Partitioning

Notice that `10` can be `partitioned` into `2 fives`

$$
10 = 5 \cdot 2
$$

If we move on to the `next power of 10`, we multiply by `10` to get this

$$
100 = 5 \cdot 20
$$

Noticing the `10` in the partition, we can split it like how we did before

$$
1000 = 5 \cdot 200
$$


Thus, we get this pattern that replicates itself

$$
\begin{align}
    10 &= 5 \cdot 2 \\
    100 &= 5 \cdot 20 \\
    1000 &= 5 \cdot 200 \\
    &\vdots \\
\end{align}
$$

## Generalisation of Such Partitioning

Using this pattern, where we `multiply by 10`, and split `10` into `9` and `1`
- All the `non-negative powers of 10` can be partitioned like so

$$
\boxed{
    10^{n} = 5 \biggr( 2 \cdot 10^{n-1} \biggr)
}
$$

# Factoring the Number

Going back to trying to factor any number to make it `divisible by 5`

$$
\text{Number} = 
\space \underline{k \cdot 10^{n}} + 
\underline{m \cdot 10^{n-1}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

Since we know this

$$
10^{n} = 5 \biggr( 2 \cdot 10^{n-1} \biggr)
$$

We factor it in

$$
\text{Number} = 
5 \biggr( 2 \cdot 10^{n-1} \biggr) k + 
5 \biggr( 2 \cdot 10^{n-2} \biggr) m + 
\cdots + 
5 \biggr( 2 \cdot 10^{1} \biggr) c + 
5 \biggr( 2 \cdot 10^{0} \biggr) b + 
a 
$$

Factor out the constant `5`, and `2` for making the equation simpler, and we get

$$
\boxed{
    \text{Number} = 
    5 \Biggr(
        2 \biggr(
            10^{n-1} k + 
            10^{n-2} m + 
            \cdots + 
            10^{1} \cdot c + 
            10^{0} \cdot b 
        \biggr)
    \Biggr) + 
    a 
}
$$



# Conclusion

Thus, we see that `this part` of the number is a `multiple of 5`
- Thus making it `divisible by 5`

$$
\textcolor{lightgray}{\text{Number} =} 
5 \Biggr(
    2 \biggr(
        10^{n-1} k + 
        10^{n-2} m + 
        \cdots + 
        10^{1} \cdot c + 
        10^{0} \cdot b 
    \biggr)
\Biggr) \textcolor{lightgray}{+ 
a}
$$

Thus for the `whole number to be divisible by 9`,  this part also has to be a `multiple of 9`

$$
\textcolor{lightgray}{\text{Number} = 
5 \Biggr(
    2 \biggr(
        10^{n-1} k + 
        10^{n-2} m + 
        \cdots + 
        10^{1} \cdot c + 
        10^{0} \cdot b 
    \biggr)
\Biggr)} + 
a
$$

Realise that it is the `digit in the ones place`

$$
\textcolor{lightgray}{
    \text{Number} \space = \space} 
\overline{\textcolor{lightgray}{k \space m \space \cdots \space c \space b} \space a}
$$

Thus, to determine if the whole number is `divisible by 5`,  the digit in the `ones place` has to be a `multiple of 5`


There are only `2 digits` that is a `multiple of 5`
- `0` itself, as it is a multiple of all numbers, and the `0-th multiple of 5`
- and  `5`, the `first multiple of 5`

All the other multiples of 5 are `too big` to be a `positive single digit`

$$
\text{All digits that are a multiple of 5:} \quad 0, 5
$$

# Divisibility by 5 Rule

Thus, to determine if a number is `divisible by 5`

$$
\boxed{
    \text{ The digit in the ones place has to be 0 or 5}
}
$$
