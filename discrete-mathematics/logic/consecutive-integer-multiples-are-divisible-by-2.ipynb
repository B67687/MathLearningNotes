# Consecutive Integer Multiples

$$
k(k+1)(k+2)(k+3) \cdots \quad \text{where k is an integer}
$$

Given that the expressions are based in integers

> Since odd and even numbers differ by 1 everytime
> - As long as any two expressions differ by an odd number, $2n - 1$, where n $\in \mathbb{Z}$ 
> - Exactly one of the 2 numbers is **even**, and the other must be **odd**
