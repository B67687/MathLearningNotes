# 三角形数

我们知道自然数的等差求和公式为：

$$
\sum_{k=1}^{n} k = 1 + 2 + 3 + \cdots + n = \dfrac{n(n+1)}{2}
$$

让我们来尝试一些递归性的操作！

我们来构造一个**从 1 到每一点的等差求和序列**：

$$
\sum_{k=1}^{1} k \quad , \space \sum_{k=1}^{2} k \quad , \space \sum_{k=1}^{3} k \quad , \space \cdots , \sum_{k=1}^{n} k
$$

应用这个公式：

$$
\dfrac{\boxed{1}(\boxed{1}+1)}{2}, \dfrac{\boxed{2}(\boxed{2}+1)}{2}, \dfrac{\boxed{3}(\boxed{3}+1)}{2}, \cdots, \dfrac{n(n+1)}{2}
$$

代入计算后得到：

$$
1, 3, 6, 10, \cdots, \dfrac{n(n+1)}{2}
$$

这些数其实就是所谓的 **三角形数**！

它们表示用点排列成一个**等边三角形**时所需的点数。

例如：
- 第 1 个三角形数是 `1` → 一个点

$$
\bullet
$$

- 第 2 个是 `3` → 排成两行的三角形 `（1 + 2）`

$$
\begin{array}{c}
    \bullet \\
    \bullet \quad \bullet \quad \bullet \\
\end{array}
$$

- 第 3 个是 `6` → 排成三行的三角形 `（1 + 2 + 3）`

$$
\begin{array}{c}
    \bullet \\
    \bullet \quad \bullet \quad \bullet \\
    \bullet \quad \bullet \quad \bullet \quad \bullet \quad \bullet \\
\end{array}
$$


- 第 4 个是 `10` → 排成四行的三角形 `（1 + 2 + 3 + 4）`

$$
\begin{array}{c}
    \bullet \\
    \bullet \quad \bullet \quad \bullet \\
    \bullet \quad \bullet \quad \bullet \quad \bullet \quad \bullet \\
    \bullet \quad \bullet \quad \bullet \quad \bullet \quad \bullet \quad \bullet \quad \bullet \\
\end{array}
$$

依此类推。

因此，第 `n` 个三角形数就是前 `n` 个自然数的和：

$$
T_n = \sum_{k=1}^n k = \dfrac{n(n+1)}{2}
$$

这些数描述了可以用点构成不断扩大的等边三角形的数量模式。

# 三角形数之和

既然我们已经了解了什么是三角形数，接下来我们就可以来求它的和。这个和的形式如下：

$$
\dfrac{\boxed{1} \left( \boxed{1} + 1 \right)}{2} + \dfrac{\boxed{2} \left( \boxed{2} + 1 \right)}{2} + \dfrac{\boxed{3} \left( \boxed{3} + 1 \right)}{2} + \cdots + \dfrac{\boxed{n} \biggr( \boxed{n} + 1 \biggr)}{2}
$$

将其压缩表示为求和形式就是：

$$
\sum_{k=1}^n \dfrac{k(k+1)}{2}
$$

当我们把常数提出时，它变为：

$$
\dfrac{1}{2} \sum_{k=1}^n k(k+1)
$$

注意这是**连续两个整数乘积**的和：

$$
\textcolor{lightgray}{\dfrac{1}{2}} \sum_{k=1}^n k(k+1)
$$

我们在另一个笔记中已经推导过这个和，结果是：

$$
\sum_{k=1}^n k(k+1) = \dfrac{n(n+1)(n+2)}{3}
$$

因此，代入后我们有：

$$
\sum_{k=1}^n k(k+1) = \dfrac{1}{2} \cdot \boxed{\dfrac{n(n+1)(n+2)}{3}}
$$

---

再将常数合并，我们就得到了通用公式：

$$
\boxed{
    \sum_{k=1}^n k(k+1) = \dfrac{n(n+1)(n+2)}{6}
}
$$
