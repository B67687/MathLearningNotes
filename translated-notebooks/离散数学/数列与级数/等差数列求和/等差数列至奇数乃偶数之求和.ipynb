# 自然数直到某个奇数或偶数的和是多少？

我们都知道自然数的求和公式：

$$
1 + 2 + \cdots + (n-1) + n
$$

它的结果是：

$$
\dfrac{n(n+1)}{2}
$$

---

但如果我们要求的是直到某个**奇数**或**偶数**的自然数之和呢？

---

我们知道，当 `n` 是奇数/偶数序列中的第 `n` 项时，

- **奇数**可以用一个**对称、长度为奇数的和**来表示：

$$
1 + 2 + \cdots + n + \cdots + 2 + 1
$$

- **偶数**可以用一个**对称、长度为偶数的和**来表示：

$$
1 + 2 + \cdots + n + n + \cdots + 2 + 1
$$

这两个“对称和”之间的唯一区别是：**偶数和会多出一项 `n`**，因为它是偶数长度的对称。

# 求到第 `n` 个奇数的和

构造从 `1` 到第 `n` 个奇数的自然数之和，如下所示：

$$
1 + 2 + 3 + 4 + 5 + \cdots + \underline{\text{第 n 个奇数}}
$$

我们可以把它拆分成一个**奇数列**和一个**偶数列**：

$$
\begin{array}
    & +1 & & +3 & & +5 & \cdots & + \underline{\text{第 (n-1) 个奇数}} && + \underline{\text{第 n 个奇数}} \\
    & +2 & & +4 & & +6 & \cdots & + \underline{\text{第 (n-1) 个偶数}} \\
\end{array}
$$

如果用对称和的方式来观察，它看起来像这样：

$$
\begin{array}
    & \quad \space \space 1 + 2 + \cdots + (n-1) + n + (n-1) + \cdots + 2 + 1 \\
    \space + 1 + 2 + \cdots + (n-1) \qquad +
    (n-1) + \cdots + 2 + 1 \\
\end{array}
$$

现在我们把它们加在一起：

$$
\begin{align}
    & \text{直到第 n 个奇数的自然数和} \\
    &= 2 \biggr( 1 + 2 + \cdots + (n-1) \biggr) + n + 2 \biggr( (n-1) + \cdots + 2 + 1 \biggr) \\
\end{align}
$$

展开并化简：

$$
4 \biggr( 1 + 2 + \cdots + (n-1) \biggr) + n
$$

合并两个相同的和：

$$
4 \biggr( \dfrac{n(n-1)}{2} \biggr) + n 
$$

进一步化简：

$$
\begin{align}
    & 2n(n-1) + n \\
    \\
    &= 2n^{2} -2n + n \\
    \\
    &= 2n^{2} - n \\
    \\
    &= n (2n - 1)
\end{align}
$$

因此，

$$
\boxed{
    \begin{align}
        & \text{直到第 n 个奇数的自然数和} \\
        &= n (2n - 1)
    \end{align}
}
$$

# 求到第 `n` 个偶数的和

构造从 `1` 到第 `n` 个偶数的自然数之和：

$$
1 + 2 + 3 + 4 + 5 + \cdots + \underline{\text{第 n 个偶数}}
$$

同样地，它可以拆分为奇数列和偶数列：

$$
\begin{array}
    & +1 & & +3 & & +5 & \cdots & \underline{\text{第 (n-1) 个奇数}} &  & + \underline{\text{第 n 个奇数}} \\
    & +2 & & +4 & & +6 & \cdots & + \underline{\text{第 (n-1) 个偶数}} && + \underline{\text{第 n 个偶数}} \\
\end{array}
$$

以对称和的方式来看，它是这样的：

$$
\begin{array}
    & \quad \space \space 1 + 2 + \cdots + (n-1) + n + n + (n-1) + \cdots + 2 + 1 \\
    + \space\space 1 + 2 + \cdots + (n-1) \quad  + n \quad
    + (n-1) + \cdots + 2 + 1 \\
\end{array}
$$

现在我们加在一起：

$$
\begin{align}
    & \text{直到第 n 个偶数的自然数和} \\
    &= 2 \biggr( 1 + 2 + \cdots + (n-1) \biggr) + 3n + 2 \biggr( (n-1) + \cdots + 2 + 1 \biggr) \\
\end{align}
$$

展开并化简：

$$
4 \biggr( 1 + 2 + \cdots + (n-1) \biggr) + 3n
$$

合并两个相等的部分：

$$
4 \biggr( \dfrac{n(n-1)}{2} \biggr) + 3n 
$$

继续化简：

$$
\begin{align}
    & 2n(n-1) + 3n \\
    \\
    &= 2n^{2} -2n + 3n \\
    \\
    &= 2n^{2} + n \\
    \\
    &= n (2n + 1)
\end{align}
$$

因此，

$$
\boxed{
    \begin{align}
        & \text{直到第 n 个偶数的自然数和} \\
        &= n (2n + 1)
    \end{align}
}
$$

# 结论

---

综上所述，自然数直到某个特定项的和如下：

##### 直到第 `n` 个奇数：

$$
\boxed{
    \begin{align}
        & \text{直到第 n 个奇数的自然数和} \\
        &= n (2n - 1)
    \end{align}
}
$$

##### 直到第 `n` 个偶数：

$$
\boxed{
    \begin{align}
        & \text{直到第 n 个偶数的自然数和} \\
        &= n (2n + 1)
    \end{align}
}
$$
