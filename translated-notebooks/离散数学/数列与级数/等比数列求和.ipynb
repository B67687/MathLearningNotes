# 什么是等比数列的和？

一个等比数列的和具有如下形式：

$$
S_n = a + ar + ar^{2} + ar^{3} + \cdots + ar^{n-1}, \text{ 共 n 项}
$$

其中 `a` 是首项，`r` 是每一项与前一项之间的固定倍数（公比）

这也意味着它可以表示为以下形式：

$$
S_n = a \space (1 + r + r^{2} + r^{3} + \cdots + r^{n-1})
$$

#### 如何推导等比数列和的一般公式？
- 注意到每一项与下一项的比值正好是 `r` 吗？
- 我们的目标是简化这个公式
- 这意味着每一项可以通过相同的比率 `r` 移动到下一项
- 在一个数列中，当整个序列乘以 `r` 后，中间的大多数项会保持相同
- 利用这个特性，我们可以构造一个差式，几乎消去所有中间项，只留下首项和末项


# 动手推导

我们先专注于这部分：

$$
\textcolor{lightgray}{S_n = a \space (} 1 + r + r^{2} + r^{3} + \cdots + r^{n-1}\textcolor{lightgray}{)}
$$

设：

$$
Z = 1 + r + r^{2} + r^{3} + \cdots + r^{n-1}
$$

现在我们构造一个差式：用它本身减去它的 `r` 倍

$$
\begin{align}
    Z \space (1-r) &= (1 + r + r^{2} + r^{3} + \cdots + r^{n-1}) \space (1-r)\\ 
    \\
    &= \space (1 + r + r^{2} + r^{3} + \cdots + r^{n-1}) \\
    & \quad - (r + r^{2} + r^{3} + r^{4} + \cdots + r^{n}) \\
    \\
    &= (1 + r + r^{2} + r^{3} + r^{4} + \cdots + r^{n-1}) \\
    &\qquad \space - (r + r^{2} + r^{3} + r^{4} + \cdots + r^{n-1} + r^{n})
\end{align}
$$

这时我们可以消去中间的大部分项，并引入一个新的最大项：

$$
\begin{align}
    Z \space (1-r) &= (1 + \cancel{r} + \cancel{r^{2}} + \cancel{r^{3}} + \cancel{r^{4}} + \cdots + \cancel{r^{n-1}}) \\
    &\qquad \space - (\cancel{r} + \cancel{r^{2}} + \cancel{r^{3}} + \cancel{r^{4}} + \cdots + \cancel{r^{n-1}} + r^{n}) \\
    &= 1 - r^{n}\\
\end{align}
$$

因此，可以得到 $Z$ 的简化公式：

$$
\boxed{
    Z = \dfrac{1 - r^{n}}{1-r}
}
$$

---

差式的方向可以互换，我们也可以这样写：

$$
\begin{align}
    Z \space (r-1) &= \qquad \quad (\cancel{r} + \cancel{r^{2}} + \cancel{r^{3}} + \cancel{r^{4}} + \cdots + \cancel{r^{n-1}} + r^{n}) \\
    &\qquad - (1 + \cancel{r} + \cancel{r^{2}} + \cancel{r^{3}} + \cancel{r^{4}} + \cdots + \cancel{r^{n-1}}) \\
    \\
    &= r^{n} - 1\\
\end{align}
$$

于是我们得到了另一个形式的公式：

$$
\boxed{
    Z = \dfrac{r^{n} - 1}{r-1}
}
$$

---

我们其实可以在这里停下来，因为这已经描述了最基本的等比数列：

$$
\boxed{
    Z = \dfrac{1 - r^{n}}{1-r} \qquad \text{或} \qquad  Z = \dfrac{r^{n} - 1}{r-1}
}
$$

但由于常数也是公式的一部分，我们继续代入完整公式

# 结论

将其代回原式：

$$
\textcolor{lightgray}{S_n = a \space (}1 + r + r^{2} + r^{3} + \cdots + r^{n-1} \textcolor{lightgray}{)}
$$

我们得到：

$$
S_n = a \left( \dfrac{1 - r^{n}}{1-r} \right) \qquad \text{或} \qquad S_n = a \left( \dfrac{r^{n} - 1}{r-1} \right)
$$

更常见的写法是：

$$
\boxed{
    S_n = \dfrac{a(1 - r^{n})}{1-r} \qquad \text{或} \qquad S_n = \dfrac{a(r^{n} - 1)}{r-1}
}
$$


# 指数行为

对于指数函数来说，如果 $|r| > 1$，则 $r^{n}$ 会呈指数增长，趋向于无穷大  
- 如果 $|r| < 1$，则 $r^{n}$ 会呈指数衰减，最终趋于有界

因此，当 $|r| > 1$ 时：

$$
\lim_{n \to \infty}{r^n} = \infty
$$

同样地，对等比数列也有：

$$
\lim_{n \to \infty}{\dfrac{a(1-r^n)}{1-r}} = \infty
$$

所以对于 $|r| > 1$，该等比数列没有收敛结果

那如果是 $|r| < 1$ 呢？

$$
\lim_{n \to \infty}{r^n} = 0
$$

此时等比数列变为：

$$
\lim_{n \to \infty}{\dfrac{a(1-r^n)}{1-r}} 
$$

除了极限部分，其余都被视为常数：

$$
\dfrac{a \biggr( 1- (\lim_{n \to \infty}{r^n}) \biggr)}{1-r} 
$$

代入极限：

$$
\dfrac{a(1 - 0)}{1-r} 
$$

化简得：

$$
\dfrac{a}{1-r} 
$$

因此：

$$
\lim_{n \to \infty}{\dfrac{a(1-r^n)}{1-r}} = \dfrac{a}{1-r}
$$

# 收敛形式

由此我们得出结论：当 $|r| < 1$ 时，等比数列在趋于无穷时存在一个收敛结果！

$$
\boxed{
    \text{当 } |r| < 1 \text{ 时的无穷等比数列和} = \dfrac{a}{1-r}
} 
$$
