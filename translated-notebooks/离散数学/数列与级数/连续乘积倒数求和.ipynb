# 连续乘积的倒数级数

考虑这个级数：

$$
\dfrac{1}{2} + \dfrac{1}{6} + \dfrac{1}{12} + \dfrac{1}{20} + \dfrac{1}{30} + \cdots
$$

注意分母是由连续的乘积组成的：

$$
\dfrac{1}{1\cdot2} + \dfrac{1}{2\cdot3} + \dfrac{1}{3\cdot4} + \dfrac{1}{4\cdot5} + \dfrac{1}{5\cdot6} + \cdots
$$

它可以简化为：

$$
\sum_{k=1}^n \dfrac{1}{k(k+1)}
$$

我们如何求解这个级数？
- 我们可以通过将分数拆分成更小的部分（每个部分对应一个质因子的分母）来进一步简化它。

# 使用部分分式分解求解

聚焦于这个表达式：

$$
\textcolor{lightgray}{\sum_{k=1}^n} \dfrac{1}{k(k+1)}
$$

设置部分分式分解：

$$
\dfrac{1}{k(k+1)} = \dfrac{A}{k} + \dfrac{B}{k+1}
$$

开始求解：

令 `A` 和 `B` 使得两边分母相同：

$$
\dfrac{1}{k(k+1)} = \dfrac{A(k+1)}{k(k+1)} + \dfrac{Bk}{(k+1)k}
$$

合并后得到：

$$
\dfrac{1}{k(k+1)} = \dfrac{A(k+1) + Bk}{k(k+1)}
$$

由于分母相同，我们可以只比较分子：

$$
1 = A(k+1) + Bk
$$

展开并按 `k` 合并项：

$$
1 = Ak + A + Bk
$$

即：

$$
1 = (A+B)k + A
$$

比较常数项系数，得到：

$$
\boxed{A = 1}
$$

比较 `k` 的系数，得到：

$$
A + B = 0
$$

代入 `A = 1` 得到：

$$
1 + B = 0 \quad \Rightarrow \quad \boxed{B = -1}
$$

因此，我们完成部分分式分解如下：

$$
\dfrac{1}{k(k+1)} = \dfrac{\boxed{1}}{k} + \dfrac{\boxed{-1}}{k+1}
$$

重新整理为：

$$
\boxed{
    \dfrac{1}{k(k+1)} = \dfrac{1}{k} - \dfrac{1}{k+1}
}
$$


# 折叠和法

回到原来的求和：

$$
\sum_{k=1}^n \dfrac{1}{k(k+1)} = \sum_{k=1}^n \left( \dfrac{1}{k} - \dfrac{1}{k+1} \right)
$$

你会发现这是一个**折叠和**（telescoping sum），意味着中间的许多项会相互抵消，从而留下一个简洁的结果！

让我们把它展开看看：

$$
\begin{align*}
    \sum_{k=1}^n \left( \dfrac{1}{k} - \dfrac{1}{k+1} \right) &= \dfrac{1}{1} - \dfrac{1}{2} \\
    & + \left( \dfrac{1}{2} - \dfrac{1}{3} \right) \\
    & + \left( \dfrac{1}{3} - \dfrac{1}{4} \right) \\
    & \vdots \\
    & + \left( \dfrac{1}{n} - \dfrac{1}{n+1} \right)
\end{align*}
$$

然后我们抵消掉中间相同的正负项：

$$
\begin{align*}
    \sum_{k=1}^n \left( \dfrac{1}{k} - \dfrac{1}{k+1} \right) &= \dfrac{1}{1} - \cancel{\dfrac{1}{2}} \\
    & + \left( \cancel{\dfrac{1}{2}} - \cancel{\dfrac{1}{3}} \right) \\
    & + \left( \cancel{\dfrac{1}{3}} - \cancel{\dfrac{1}{4}} \right) \\
    & + \left( \cancel{\dfrac{1}{4}} - \cdots \right) \\
    & \vdots \\
    & + \left( \cdots - \cancel{\dfrac{1}{n}} \right) \\
    & + \left( \cancel{\dfrac{1}{n}} - \dfrac{1}{n+1} \right) \\
    & = 1 - \dfrac{1}{n+1}
\end{align*}
$$

所以我们得到：

$$
\sum_{k=1}^n \left( \dfrac{1}{k} - \dfrac{1}{k+1} \right) = 1 - \dfrac{1}{n+1}
$$

继续化简这部分：

$$
1 - \dfrac{1}{n+1} = \dfrac{n+1}{n+1} - \dfrac{1}{n+1} = \dfrac{(n+1) - 1}{n+1} = \dfrac{n}{n+1}
$$

因此我们有：

$$
\sum_{k=1}^n \left( \dfrac{1}{k} - \dfrac{1}{k+1} \right) = \dfrac{n}{n+1}
$$

# 结论

因此，连续乘积倒数的和为：

$$
\boxed{
    \sum_{k=1}^n \dfrac{1}{k(k+1)} = \dfrac{n}{n+1}
}
$$
