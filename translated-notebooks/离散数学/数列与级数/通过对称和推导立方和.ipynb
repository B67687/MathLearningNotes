# 通过对称求和推导立方和公式

我们从立方和开始：

$$
\sum_{k=1}^n k^3 = 1^3 + 2^3 + \cdots + n^3
$$

注意观察我们可以将每个立方项写成“平方乘以它本身”的形式：

$$
\sum_{k=1}^n k \cdot k^2 = 1 \cdot 1^2 + 2 \cdot 2^2 + \cdots + n \cdot n^2
$$

于是我们发现每一行都包含 `n` 个该数的平方：

$$
\sum_{k=1}^n k^3 =
\begin{array}{c|c|cc|ccc|cccc}
    & 1 & 2 & 2 & 3 & 3 & 3 & n & \cdots & n \\
    & & 2 & 2 & 3 & 3 & 3 & n & \cdots & n \\
    & & & & 3 & 3 & 3 & n & \cdots & n \\ 
    & & & & & & & \vdots & & \vdots \\
    & & & & & & & n &\cdots & n \\
\end{array}
$$

我们也知道，平方数其实就等于前 `n` 个奇数之和：

$$
\sum_{k=1}^n k^3 =
\begin{array}{c|c|ccc|ccccc|c|cccccc}
    & 1 & & 2 & & & & 3 & & & & & & & & n \\
    & & 2 & 2 & 2 & & 3 & 3 & 3 & & \cdots & & & & n & n & n \\
    & & & & & 3 & 3 & 3 & 3 & 3 & & & & n & n & n & n & n \\
    & & & & & & & & & & & & & & & \vdots & & & \\
    & & & & & & & & & & & n & \cdots & n & n & n & n & n & \cdots & n\\
\end{array}
$$

如果我们按列来观察，这些其实是具有**对称结构**的求和形式：

$$
\sum_{k=1}^n k^3 =
\begin{array}{c|c|ccc|ccccc|c|cccccc}
    \\
    & 1 & 2 & 4 & 2 & 3 & 6 & 9 & 6 & 3 & \cdots & n & 2n & 3n & \cdots & n^{2} & \cdots & 3n & 2n & n \\
    \\
\end{array}
$$

而这些本质上都是一个简单对称数列 `1 + 2 + 3 + ... + n + ... + 3 + 2 + 1` 的倍数形式。

$$
\sum_{k=1}^n k^3 =
\begin{array}{cc|c|ccc|ccccc|c|ccccc}
    & \text{Multiple} & 1 & & 2 & & & & 3 & & & \cdots & & & & & n & & & & & \\
    & \hline
    & 1 & 1 & 2 & 1 & 1 & 2 & 3 & 2 & 1 & \cdots & 1 & 2 & 3 & \cdots & n & \cdots & 3 & 2 & 1 \\
    \\
\end{array}
$$

# 对称性与平方数

你也会发现：

- 这个数列具有**对称结构**
- 对称求和项数是**奇数个**
- 奇数个项的对称和恰好构成一个**平方数**，因为每新增一个奇数项，相当于给当前正方形增加了一个“右下角”

$$
\colorbox{lightgray}{因此，每一个新的“奇数长度对称和”都在扩展这个正方形的右下角}
$$

- 表中 “倍数” 一列表示贡献到该“右下角”的倍数

# 结论

我们发现：

$$
\sum_{k=1}^n k^3 = (1 + 2 + 3 + \cdots + n) \quad \text{的 } (1 + 2 + 3 + \cdots + n) \text{ 份}
$$

也就是：

$$
\sum_{k=1}^n k^3 = (1 + 2 + 3 + \cdots + n) \quad \times \quad (1 + 2 + 3 + \cdots + n)
$$

即：

$$
\sum_{k=1}^n k^3 = (1 + 2 + 3 + \cdots + n)^2
$$

代入自然数和公式后得出：

$$
\boxed{
    \sum_{k=1}^n k^3 = \left[ \dfrac{n(n+1)}{2} \right]^2
}
$$
