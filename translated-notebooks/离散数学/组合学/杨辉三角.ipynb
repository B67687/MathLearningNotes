# 杨辉三角

这是杨辉三角：

$$
\begin{array}
    & & & & & & 1 & & & & & &\\
    & & & & 1 & & 1 & & & &\\
    & & & 1 & & 2 & & 1 & & &\\
    & & 1 & & 3 & & 3 & & 1 & &\\
    & 1 & & 4 & & 6 & & 4 & & 1 &\\
    1 & & 5 & & 10 & & 10 & & 5 & & 1\\
\end{array}
$$

除了 `1` 以外的每一个数，都等于它正上方两个数的和。

通常我们把 `1` 视作杨辉三角的起点，其余数字都是由它上面两个数相加得到的。  
- 对于三角形的两边来说，相当于 `1 + 0`，所以它们永远是 `1`

我们也知道它与二项式分布有关：每一行对应的是一个二项式展开中各项的系数  
- 这些系数也可以用组合数来表示，例如 ${}^{2}C_1$、${}^{4}C_3$、${}^{8}C_2$ 等等

这就引出了几个问题：
- 是否也可以从组合的角度来构建杨辉三角？
- 是否能解释出为什么会有 `1` 出现？
- 它与我们所熟知的杨辉三角的加法性质之间有什么联系？

让我们一起来探索！

# 选择的类型

为了更清楚地理解这里的“选择”是什么意思，我们可以定义如下：

$$
\dbinom{n}{k}, \text{其中 } n \text{ 是总数，} k \text{ 是我们要从中选出的数量}
$$

---

#### **选择负数个东西** —— 不可能  
- 选择的本质是从一堆 `n` 个物品中选出 `k` 个进行组合。因此，选比零还少的东西是不可能的。

#### **选择零个东西** —— 只有 1 种方式  
- 唯一能做的事就是什么都不选。

#### **从 1 到 n 中选择** —— 有若干种方式  
- 大于 1 的数量。

#### **选择所有东西** —— 只有 1 种方式  
- 唯一能做的事就是全部选上。

#### **选择比总数还多的东西** —— 不可能  
- 因为你不可能选出比你拥有的还要多的东西。

---

# 重建杨辉三角

现在我们了解了这些基本规律，我们可以从头开始重新构建杨辉三角。

一开始我们什么都没有，所以我们只能选择“没有东西”：

$$
\dbinom{0}{0} = 1
$$

当我们拥有 1 个东西时，我们就多了两个选择：可以选择 `0` 个或 `1` 个  
- 总共两种选择

$$
\begin{array}{ccccccc}
    & \dbinom{0}{0} = 1 &\\
    \dbinom{1}{0} = 1 & & \dbinom{1}{1} = 1\\ 
\end{array}
$$

我们可以继续这个选择过程，不断扩展下去：

$$
\begin{array}{ccccccc}
    & & & \dbinom{0}{0} = 1 & & & \\
    & & \dbinom{1}{0} = 1 & & \dbinom{1}{1} = 1 & &\\ 
    & \dbinom{2}{0} = 1 & & \dbinom{2}{1} = 2 & & \dbinom{2}{2} = 1 &\\
    \dbinom{3}{0} = 1 & & \dbinom{3}{1} = 3 & & \dbinom{3}{2} = 3 & & \dbinom{3}{3} = 1
\end{array}
$$

如何计算这些选择的具体数值，请参考排列组合章节。


# 组合与加法的关系？

但是组合和加法之间有什么关系呢？

为什么会有：

$$
\dbinom{0}{0} = \dbinom{1}{0} = \dbinom{1}{1} = \dbinom{n}{0} = \dbinom{n}{1}
$$

还有：

$$
\dbinom{1}{0} + \dbinom{1}{1} = \dbinom{2}{1}, \quad \dbinom{2}{1} + \dbinom{2}{2} = \dbinom{3}{2}, \quad \text{依此类推}
$$

看起来似乎没什么关联？  
其实——大有关系！

当我们增加一个新元素到总数中时，对于这个新元素我们有两种选择：要么选它，要么不选它。


# 示例讲解

---

## 从 7 选 3 到 8 选 3

如果我们原本有 7 个元素，现在加入第 8 个元素，变成 8 个元素，我们要从中选 3 个：

- 第 8 个元素要么被选中，要么不被选中。

于是，“8 选 3”包含以下两种情况：

- 如果被选中：

$$
\text{(第 8 个被选中), (剩下的 7 个中还需再选 2 个)} \longrightarrow \dbinom{1}{1} \space \text{然后} \space \dbinom{7}{2} \longrightarrow \dbinom{1}{1} \cdot \dbinom{7}{2}
$$

- 如果未被选中：

$$
\text{(第 8 个未被选中), (剩下的 7 个中仍需选 3 个)} \longrightarrow \dbinom{1}{0} \space \text{然后} \space \dbinom{7}{3} \longrightarrow \dbinom{1}{0} \cdot \dbinom{7}{3}
$$

因此，“8 选 3”可以写成这两个情况之和：

$$
\underline{
    \dbinom{8}{3} = \dbinom{1}{1} \dbinom{7}{2} + \dbinom{1}{0} \dbinom{7}{3}
}
$$

进一步简化为：

$$
\underline{
    \dbinom{8}{3} = \dbinom{7}{2} + \dbinom{7}{3}
}
$$

因为 $\dbinom{1}{1} = 1$，而 $\dbinom{1}{0} = 1$

---

如果我们选择零个元素呢？


## 8 选 0 的情况

假设我们从 “7 选 0” 开始，现在加入第 8 个元素，变成 “8 选 0”

- 显然，第 8 个元素不能被选中，因为我们什么都不能选

$$
0
$$

- 所以我们只能从剩下的 7 个中选择 0 个，这只有 1 种方法：

$$
\dbinom{7}{0} = 1
$$

因此：

$$
\dbinom{8}{0} = 0 + \dbinom{7}{0} = 1
$$

即：

$$
\underline{\dbinom{8}{0} = 1}
$$

虽然我们知道任何数“选 0”都是 1，但我们通过这个逻辑验证了这一结论在整个过程中依然成立。

甚至对于负整数，“选 0”也等于 1：

$$
\underline{\dbinom{n}{0}}, \space \text{其中 } n \in \mathbf{Z}
$$


如果我们想选择所有元素呢？这又该如何用这种方式解释？

---

## 8 选 8 的情况

我们从 “7 选 8” 开始，加入第 8 个元素后变为 “8 选 8”

- 第 8 个元素要么被选中，要么不被选中：

- 如果被选中：

$$
\text{(第 8 个被选中), (剩下的 7 个中还需再选 7 个)} \longrightarrow \dbinom{1}{1} \cdot \dbinom{7}{7}
$$

- 如果未被选中：

$$
\text{(第 8 个未被选中), (剩下的 7 个中仍需选 8 个)} \longrightarrow \dbinom{1}{0} \cdot \dbinom{7}{8}
$$

于是：

$$
\dbinom{8}{8} = \dbinom{1}{1} \dbinom{7}{7} + \dbinom{1}{0} \dbinom{7}{8}
$$

- 我们知道 $\dbinom{7}{7} = 1$，因为全选只有一种方法  
- 而 $\dbinom{7}{8} = 0$，因为你不能从 7 个中选 8 个

因此：

$$
\begin{align}
    \dbinom{8}{8} &= \dbinom{1}{1} \dbinom{7}{7} + \dbinom{1}{0} \dbinom{7}{8} \\
    & = \dbinom{1}{1} (1) + \dbinom{1}{0} (0) \\
    & = 1(1) + 1(0) \\
    & = 1 \\
\end{align}
$$

我们可以把这个结论推广到任意非负整数 $ n $，都有：

$$
\underline{\dbinom{n}{n} = 1}, \space \text{其中 } n \in \mathbf{Z}_{\ge 0}
$$

这样我们就通过递归的方式解释了为什么“n 选 n”恒等于 1。

# 选择的加法通用公式

我们可以将增加总元素数量的过程一般化，设 `n` 是总元素数，我们从中选 `m` 个：

从 $\dbinom{n}{m}$ 到 $\dbinom{n+1}{m}$，我们有：

$$
\dbinom{n}{m} \longrightarrow \dbinom{n+1}{m} = \underline{\dbinom{1}{1} \dbinom{n}{m-1} + \dbinom{1}{0} \dbinom{n}{m}} = \underline{\dbinom{n}{m-1} + \dbinom{n}{m}} 
$$

我们可以进一步推广到任意 $\dbinom{n}{m}$，有：

$$
\begin{align}
    \dbinom{n}{m} & = \dbinom{1}{1} \dbinom{n-1}{m-1} + \dbinom{1}{0} \dbinom{n-1}{m} \\
    & = \textcolor{lightgray}{(1)} \dbinom{n-1}{m-1} + \textcolor{lightgray}{(1)} \dbinom{n-1}{m} \\
    & = \dbinom{n-1}{m-1} + \dbinom{n-1}{m}
\end{align}
$$

于是我们从：

$$
\boxed{
    \dbinom{n}{m} = \dbinom{1}{1} \dbinom{n-1}{m-1} + \dbinom{1}{0} \dbinom{n-1}{m}
}
$$

得出：

$$
\underline{
    \dbinom{n}{m} = \dbinom{n-1}{m-1} + \dbinom{n-1}{m}
}
$$

# 测试这个公式

我们来测试一下极端情况：

- 当 `n 选 0`：

$$
\begin{align}
    \dbinom{n}{0} & = \dbinom{1}{1} \dbinom{n-1}{-1} + \dbinom{1}{0} \dbinom{n-1}{0} \\
    & = \dbinom{n-1}{-1} + \dbinom{n-1}{0} \\
    & = 0 + 1 \\
    & = 1
\end{align}
$$

成功！

- 当 `n 选 n`：

$$
\begin{align}
    \dbinom{n}{n} & = \dbinom{1}{1} \dbinom{n-1}{n-1} + \dbinom{1}{0} \dbinom{n-1}{n} \\
    & = \dbinom{n-1}{n-1} + \dbinom{n-1}{n} \\
    & = 1 + 0 \\
    & = 1
\end{align}
$$

也成功！

---



## 结论

总结一下，我们从“选择”的角度出发，构建了一个组合树：

$$
\begin{array}{ccccccc}
    & & & \dbinom{0}{0} = 1 & & & \\
    & & \dbinom{1}{0} = 1 & & \dbinom{1}{1} = 1 & & \\ 
    & \dbinom{2}{0} = 1 & & \dbinom{2}{1} = 2 & & \dbinom{2}{2} = 1 & \\
    \dbinom{3}{0} = 1 & & \dbinom{3}{1} = 3 & & \dbinom{3}{2} = 3 & & \dbinom{3}{3} = 1 \\
    &&& \vdots &&&
\end{array}
$$

并通过加法原理：

$$
\dbinom{n}{m} \longrightarrow \dbinom{n+1}{m} = \boxed{\dbinom{1}{1} \dbinom{n}{m-1} + \dbinom{1}{0} \dbinom{n}{m}} = \underline{\dbinom{n}{m-1} + \dbinom{n}{m}}
$$

我们得到了下一个数字是由它上面两个数字相加而来。

- 最终我们得到了简化的形式，也就是杨辉三角：

$$
\begin{array}
    & & & & & & 1 & & & & & &\\
    & & & & 1 & & 1 & & & &\\
    & & & 1 & & 2 & & 1 & & &\\
    & & 1 & & 3 & & 3 & & 1 & &\\
    & 1 & & 4 & & 6 & & 4 & & 1 &\\
    1 & & 5 & & 10 & & 10 & & 5 & & 1\\
    &&&&& \vdots &&&&&
\end{array}
$$

---

### 备注

- 表面上看，杨辉三角似乎只是简单地以 `1` 为起点，然后靠加法生成其他数字
- 但我们现在知道，它的背后隐藏着组合选择的深层逻辑！

# 总选择数（Total choices）

由于每个元素只有两种状态：“被选中”或“未被选中”，所以总的组合数为：

$$
\underbrace{\underline{2} \times \underline{2} \times \underline{2} \times \cdots \times \underline{2}}_{\text{n 次选择}}
$$

因此，总的选择数为：

$$
\underline{2^{n}}
$$

也就是说：

$$
\boxed{
    \dbinom{n}{0} + \dbinom{n}{1} + \dbinom{n}{2} + \cdots + \dbinom{n}{n} = 2^{n}
}
$$
