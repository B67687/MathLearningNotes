# 数字的构成

这是数字的基本结构：

$$
\text{数字} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n 位}} =
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

我们来看看 `10 的幂`

# 分解 10 的幂

让我们从 $10^0$ 开始  
- 它等于 `1`

$$
10^{0} = 1
$$

我们无法对它进行“分割”或“因式分解”，所以继续看 $10^1$

## 故意进行分割

注意，`10` 可以被“分割”成一个只包含 `2` 的部分：

$$
\boxed{10 = \underline{2 \cdot 5}}
$$

如果我们继续到下一个 10 的幂，乘以 `10` 得到：

$$
\boxed{100 = \underline{2 \cdot 50}}
$$

同样地，当我们继续到下一个幂时，我们有：

$$
\boxed{1000 = \underline{2 \cdot 500}}
$$

## 此种分割方式的一般形式

因此，我们得到了一个从 `第一个幂开始` 到某个 `n` 的自我重复模式：

$$
\begin{align}
    10 &= 2 \cdot 5 \\
    100 &= 2 \cdot 50 \\
    1000 &= 2 \cdot 500 \\
    &\vdots \\
    10^{n} &= 2 \biggr( 5 \cdot 10^{n-1} \biggr)
\end{align}
$$

# 按照数字结构分析

让我们再来看一遍这个数字的结构：

$$
\text{数字} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n 位}} = 
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

## 按位判断是否能被整除

由于我们知道从 `第一个幂开始` 的所有 `10 的幂` 都是 `2 的倍数`

$$
\text{数字} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space \textcolor{lightgray}{a}}}_{\text{n 位}} = 
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} \textcolor{lightgray}{+ 
\underline{a \cdot 10^{0}}}
$$

我们只需要检查 `个位数` 是否能被 `2` 整除即可：

$$
\text{数字} = \underbrace{\overline{\textcolor{lightgray}{k \space m \space \cdots \space c \space b} \space a}}_{\text{n 位}} = 
\textcolor{lightgray}{\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}}} + 
\underline{a \cdot 10^{0}}
$$

# 2 的倍数

只有 `五个数字` 是 `2 的倍数`：

$$
\begin{align}
    0 &: \space \text{它是所有数的 0 倍} \\
    2 &: \space \text{2 的第一倍} \\
    4 &: \space \text{2 的第二倍} \\
    6 &: \space \text{2 的第三倍} \\
    8 &: \space \text{2 的第四倍} \\
\end{align}
$$

其它更大的 `2 的倍数` 都太大了，不能作为正的个位数字存在。

# 被2整除的规则

因此，判断一个数字是否能被 `2` 整除：

$$
\textbf{个位上的数字必须是以下之一：0, \ 2, \ 4, \ 6, \ 8}
$$

换句话说：

> **如果个位数是偶数，则该数字能被 2 整除**

### **我们尝试用因式分解的方式看看是否还有其他方法可以判断**

# 对数字进行因式分解

我们再来看一遍这个数字的结构：

$$
\text{数字} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n 位}} = 
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

由于我们知道这个等式成立：

$$
10^{n} = 2 \biggr( 5 \cdot 10^{n-1} \biggr)
$$

我们将它代入原式中：

$$
\text{数字} = 
2 \biggr( 5 \cdot 10^{n-2} \biggr) k + 
2 \biggr( 5 \cdot 10^{n-3} \biggr) m + 
\cdots + 
2 \biggr( 5 \cdot 10^{1} \biggr) c + 
2 \biggr( 5 \cdot 10^{0} \biggr) b + 
a 
$$

提取出常数因子 `5` 和 `2`，使表达式更简洁，我们得到：

$$
\boxed{
    \text{数字} = 
    2 \Biggr(
        5 \biggr(
            10^{n-2} k + 
            10^{n-3} m + 
            \cdots + 
            10c + 
            b 
        \biggr)
    \Biggr) + 
    a 
}
$$

# 余数分析

因此，我们看到这部分是 `2 的倍数`
- 因此它是 `可被 2 整除的`

$$
\textcolor{lightgray}{\text{数字} =} 
2 \Biggr(
    5 \biggr(
        10^{n-2} k + 
        10^{n-3} m + 
        \cdots + 
        10c + 
        b 
    \biggr)
\Biggr) \textcolor{lightgray}{+ 
a}
$$

因此，要使整个数字 `能被 2 整除`，第二部分也必须是 `2 的倍数`

$$
\textcolor{lightgray}{\text{数字} = 
2 \Biggr(
    5 \biggr(
        10^{n-2} k + 
        10^{n-3} m + 
        \cdots + 
        10c + 
        b 
    \biggr)
\Biggr)} + 
a
$$

而这就是 `个位上的数字`

$$
\textcolor{lightgray}{
    \text{数字} \space = \space} 
\overline{\textcolor{lightgray}{k \space m \space \cdots \space c \space b} \space a}
$$


# 被2整除的规则（同样的规则）

因此，我们得出了相同的规则：

---

如果一个数字 `能被 2 整除`：

$$
\textbf{个位上的数字必须是以下之一：0, \ 2, \ 4, \ 6, \ 8}
$$

换句话说：

> **如果个位数是偶数，则该数字能被 2 整除**
