# 一个数的构成

$$
\text{数} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n位}} =
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

让我们看一下 `10 的幂`

# 10 的幂次分解

从$10^0$开始
- 等于`1`

$$
10^{0} = 1
$$

这个数无法进行`分割`或`因式分解`，所以我们继续看$10^1$

## 有意的分割

注意到`10`可以`分割`为`7`和`3`

$$
\boxed{ 
    \phantom{\big(} 
        10 = 7 + 3 
    \phantom{\big)} 
}
$$

继续到`下一个10的幂次`，我们乘以`10`得到

$$
100 = 7 \cdot 10 + 3 \cdot \boxed{10}
$$

注意到分割中未与倍数`7`分组的`10`，我们可以像之前那样拆分它

$$
\begin{align}
    100 &= 7 \cdot 10 + \underline{3 \cdot (7 + 3)} \\
    &= 7 \cdot 10 + 7 \cdot 3 + 3 \cdot 3 \\
    &= 7 \cdot 10 + 7 \cdot 3 + 3^{2} 
\end{align}
$$

因此，

$$
\boxed{
    \phantom{\big(} 
        100 = 7 \cdot 10 + 7 \cdot 3 + 3^{2} 
    \phantom{\big)}
}
$$

这个方法在下一个幂次同样适用，

$$
1000 = 7 \cdot 10^2 + 7 \cdot 3 \cdot 10 + 3^{2} \cdot \boxed{10}
$$

再次拆分`10`

$$
\begin{align}
    1000 &= 7 \cdot 10^2 + 7 \cdot 3 \cdot 10 + 3^{2} \cdot (7 + 3) \\
    &= 7 \cdot 10^2 + 7 \cdot 3 \cdot 10 + 7 \cdot 3^{2} + 3^3 \\
\end{align}
$$

得到

$$
\boxed{
    \phantom{\big(}
        1000 = 7 \cdot 10^2 + 7 \cdot 3 \cdot 10 + 7 \cdot 3^{2} + 3^3
    \phantom{\big)}
}
$$

继续下一个幂次，

$$
10000 = 7 \cdot 10^3 + 7 \cdot 3 \cdot 10^2 + 7 \cdot 3^{2} \cdot 10 + 3^3 \cdot \boxed{10}
$$

同样进行分组

$$
\begin{align}
    10000 &= 7 \cdot 10^3 + 7 \cdot 3 \cdot 10^2 + 7 \cdot 3^{2} \cdot 10 + 3^3 \cdot (7 + 3) \\
    &= 7 \cdot 10^3 + 7 \cdot 3 \cdot 10^2 + 7 \cdot 3^{2} \cdot 10 + 7 \cdot 3^3 + 3^4 \\
\end{align}
$$

于是得到

$$
\boxed{ 
    \phantom{\big(} 
        10000 = 7 \cdot 10^3 + 7 \cdot 3 \cdot 10^2 + 7 \cdot 3^{2} \cdot 10 + 7 \cdot 3^3 + 3^4
    \phantom{\big)}
}
$$

## 这种分割方式的广义化

因此，我们得到了从`1次幂`开始到某个`n`的重复模式

$$
\begin{align}
    10 &= 7 \cdot 3 \\
    100 &= 7 \cdot 10 + 7 \cdot 3 + 3^{2} \\
    1000 &= 7 \cdot 10^2 + 7 \cdot 3 \cdot 10 + 7 \cdot 3^{2} + 3^3 \\
    10000 &= 7 \cdot 10^3 + 7 \cdot 3 \cdot 10^2 + 7 \cdot 3^{2} \cdot 10 + 7 \cdot 3^3 + 3^4 \\
    &\vdots \\
    10^n &= 7 \cdot 3^{0} \cdot 10^{n-1} + 7 \cdot 3^1 \cdot 10^{n-2} + \cdots + 7 \cdot 3^{n-1} \cdot 10^0 + 3^{n} \\
\end{align}
$$

因此整体表达式为

$$
\boxed{
    \phantom{\Biggr(} 
        10^n = 7 \biggr( 3^{0} \cdot 10^{n-1} + 3^1 \cdot 10^{n-2} + \cdots + 3^{n-1} \cdot 10^0 \biggr) + 3^{n} \\
    \phantom{\Biggr)} 
}
$$

# 数字的因数分解

让我们重新审视这个数字

$$
\text{数字} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n位数}} = 
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

已知以下关系式

$$
10^n = 7 \biggr( 3^{0} \cdot 10^{n-1} + 3^1 \cdot 10^{n-2} + \cdots + 3^{n-1} \cdot 10^0 \biggr) + 3^{n}
$$

## 表达式简化

实际上，由于这部分已经是`7的倍数`

$$
\textcolor{lightgray}{10^n =}  7 \biggr( 3^{0} \cdot 10^{n-1} + 3^1 \cdot 10^{n-2} + \cdots + 3^{n-1} \cdot 10^0 \biggr) \textcolor{lightgray}{+} \textcolor{lightgray}{3^{n}}
$$

我们可以简化括号内的表达式为

> 其中$q_{n} = 3^{0} \cdot 10^{n-1} + 3^1 \cdot 10^{n-2} + \cdots + 3^{n-1} \cdot 10^0$


$$
\boxed{
    10^n = 7q_{n} + 3^n
}
$$

## 代数变换

将其因式分解代入

$$
\text{数字} = 
\space k \space \biggr( 7q_{n-1} + 3^{n-1} \biggr) + 
m \space \biggr( 7q_{n-2} + 3^{n-2} \biggr) + 
\cdots + 
c \space \biggr( 7q_{2} + 3^{2} \biggr) + 
b \space \biggr( 7q_{1} + 3^{1} \biggr) + 
a
$$

简化后得到

$$
\text{数字} = 
\boxed{ \phantom{\biggr(} 7k \cdot q_{n-1} + 3^{n-1}k \phantom{\biggr)}} +
\boxed{ \phantom{\biggr(} 7m \cdot q_{n-2} + 3^{n-2}m \phantom{\biggr)}} + 
\cdots +
\boxed{\phantom{\biggr(} 7c \cdot q_{2} + 3^{2}c \phantom{\biggr)}}  + 
\boxed{\phantom{\biggr(} 7b \cdot q_{1} + 3b \phantom{\biggr)}} + 
\boxed{\phantom{\biggr(} a \phantom{\biggr)}}
$$

重新分组如下

$$
\text{数字} = 
\Biggr( 
    7k \cdot q_{n-1} + 
    7m \cdot q_{n-2} +
    \cdots +
    7c \cdot q_{2} + 
    7b \cdot q_{1}
\Biggr) +
\biggr( 
    3^{n-1}k + 3^{n-2}m + \cdots + 3^{2}c + 3b + a
\biggr)
$$

按`7`的倍数重新分组

$$
\text{数字} = 
7 \Biggr( 
    kq_{n-1} + 
    mq_{n-2} +
    \cdots +
    cq_{2} + 
    bq_{1}
\Biggr) +
\biggr( 
    3^{n-1}k + 3^{n-2}m + \cdots + 3^{2}c + 3b + a
\biggr)
$$

## 重新表达位值

用$f_h$来表示`位值`更方便观察模式

$$
\text{数字} = 
7 \Biggr( 
    f_n q_{n-1} + 
    f_{n-1} q_{n-2} +
    \cdots +
    f_{3} q_{2} + 
    f_{2} q_{1}
\Biggr) +
\biggr( 
    3^{n-1} f_n + 3^{n-2} f_{n-1} + \cdots + 3^{2} f_{3} + 3^{1} f_{2} + 3^{0} f_{1}
\biggr)
$$

按`3的幂次`分组系数

$$
\text{数字} = 
7 \Biggr( 
    f_n q_{n-1} + 
    f_{n-1} q_{n-2} +
    \cdots +
    f_{3} q_{2} + 
    f_{2} q_{1}
\Biggr) +
    \sum_{h=1}^{n-1} 3^{h-1} f_h
$$

# 余数分析

由此可见，数字的`这部分`是`7的倍数`

$$
\textcolor{lightgray}{\text{数字} = }
    7 \Biggr( 
        f_n q_{n-1} + 
        f_{n-1} q_{n-2} +
        \cdots +
        f_{3} q_{2} + 
        f_{2} q_{1}
    \Biggr)
\textcolor{lightgray}{+}
\textcolor{lightgray}{
     \sum_{h=1}^{n-1} 3^{h-1} f_h
}
$$

因此，要使`整个数字能被7整除`，这部分也必须为`7的倍数`

$$
\textcolor{lightgray}{\text{数字} =
    7 \Biggr( 
        f_n q_{n-1} + 
        f_{n-1} q_{n-2} +
        \cdots +
        f_{3} q_{2} + 
        f_{2} q_{1}
    \Biggr)
} +
     \sum_{h=1}^{n-1} 3^{h-1} f_h
$$

这部分实际上是`3的(位数-1)次幂`乘以`对应位数值`的总和

$$
\boxed{
    \phantom{\big(}
        \sum_{h=1}^{n-1} 3^{h-1} f_h
    \phantom{\big)}
}
$$

# 可除性规则1

因此，判断整个数字是否`能被7整除`的条件是：

$$
\boxed{
    \phantom{\biggr(} 
        \underline{\textbf{各数字位值}}
        \space
        \textbf{乘以}
        \space
        \underline{
            \textbf{3的(位数-1)次幂}
        }  
        \space 
        \textbf{必须能被7整除}
    \phantom{\biggr)}
}
$$

# 更宽松的分割方式

注意到`10`可以`分割`为`7`和`3`

$$
\boxed{ 
    \phantom{\big(} 
        10 = 7 + 3
    \phantom{\big)} 
}
$$

继续到`10的下一个幂次`，我们乘以`10`得到

$$
\boxed{
    \phantom{\big(} 
        100 = 7 \cdot 10 + 3 \cdot 10
    \phantom{\big(} 
}
$$

这个方法在下一个幂次同样适用，

$$
\boxed{
    \phantom{\big(} 
        1000 = 7 \cdot 10^2 + 3 \cdot 10^2
    \phantom{\big(} 
}
$$

## 这种分割方式的推广

因此，我们得到了从`1次幂`开始到某个`n`的重复模式

$$
\begin{align}
    10 &= 7 + 3 \\ 
    100 &= 7 \cdot 10 + 3 \cdot 10 \\
    1000 &= 7 \cdot 10^2 + 3 \cdot 10^2 \\
    &\vdots \\
    10^n &= 7 \cdot 10^{n-1} + 3 \cdot 10^{n-1} \\
\end{align}
$$

因此整体表达式为

$$
\boxed{
    \phantom{\big(} 
        10^n = 7 \cdot 10^{n-1} + 3 \cdot 10^{n-1}
    \phantom{\big)} 
}
$$

# 因式分解这个数

再来看一遍这个数
$$
\text{数} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n 位}} = 
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

因为我们知道这个
- 其中 `n` 从 `1` 开始增加
$$
10^n = 7 \cdot 10^{n-1} + 3 \cdot 10^{n-1}
$$

我们将它代入因式分解中
$$
\text{数} = 
\space k \space \biggr( 7 \cdot 10^{n-2} + 3 \cdot 10^{n-2} \biggr) + 
m \space \biggr( 7 \cdot 10^{n-3} + 3 \cdot 10^{n-3} \biggr) + 
\cdots + 
c \space \biggr( 7 \cdot 10 + 3 \cdot 10 \biggr) + 
b \space \biggr( 7 + 3 \biggr) + 
a
$$

简化后，我们得到
$$
\text{数} = 
\boxed{ \phantom{\biggr(} 7 \cdot 10^{n-2}k + 3 \cdot 10^{n-2}k \phantom{\biggr)}} +
\boxed{ \phantom{\biggr(} 7 \cdot 10^{n-3}m + 3 \cdot 10^{n-3}m \phantom{\biggr)}} + 
\cdots +
\boxed{\phantom{\biggr(} 7 \cdot 10c + 3 \cdot 10c \phantom{\biggr)}}  + 
\boxed{\phantom{\biggr(} 7b + 3b  \phantom{\biggr)}} + 
\boxed{\phantom{\biggr(} a \phantom{\biggr)}}
$$

像这样重新分组
$$
\text{数} = 
\Biggr( 
    7 \cdot 10^{n-2}k + 
    7 \cdot 10^{n-3}m +
    \cdots +
    7 \cdot 10c + 
    7b
\Biggr) +
\biggr( 
    3 \cdot 10^{n-2}k + 3 \cdot 10^{n-3}m + \cdots + 3 \cdot 10c + 3b
\biggr) +
    a
$$

将这些组分别按 `7` 和 `3` 重新分组，我们得到
$$
\boxed{
    \text{数} = 
    7 \Biggr( 
        10^{n-2}k + 
        10^{n-3}m +
        \cdots +
        10c + 
        b
    \Biggr) +
    3 \biggr( 
        10^{n-2}k + 10^{n-3}m + \cdots + 10c + b
    \biggr) +
    a
}
$$

# 余数分析

因此，我们看到数的这一部分是 `7 的倍数`
- 从而使其 `可被7整除`
$$
\textcolor{lightgray}{\text{数} = }
    7 \Biggr( 
        10^{n-2}k + 
        10^{n-3}m +
        \cdots +
        10c + 
        b
    \Biggr)
\textcolor{lightgray}{+}
\textcolor{lightgray}{
    3 \biggr( 
        10^{n-2}k + 10^{n-3}m + \cdots + 10c + b
    \biggr) +
    a
}
$$

因此，要使 `整个数能被7整除`，这一部分也必须是 `7 的倍数`
$$
\textcolor{lightgray}{\text{数} =
    7 \Biggr( 
        10^{n-2}k + 
        10^{n-3}m +
        \cdots +
        10c + 
        b
    \Biggr)
} +
   3 \biggr( 
        10^{n-2}k + 10^{n-3}m + \cdots + 10c + b
    \biggr) +
    a
$$

意识到它包含 `3` 倍于数本身的所有数字直到 `第 n-1 位` 加上 `a`

$$
\boxed{
    \phantom{\big(}
        3 \cdot \underbrace{\overline{k \space m \space \cdots \space c \space b \space}}_{\text{n-1 位}} + a
    \phantom{\big)}
}
$$

# 可除性规则2

因此，判断`整个数字能否被7整除`的条件是：

$$
\boxed{
    \phantom{\biggr(} 
        \underline{
            \textbf{去掉个位数后剩余数字的3倍}
        } 
        \space 
        \textbf{加上} 
        \space 
        \underline{\textbf{个位数字}} 
        \space 
        \textbf{，所得结果必须能被7整除}
    \phantom{\biggr)}
}
$$

# 寻找更好的规则

如果我们让 `a` 被集成到 `3的倍数` 中，使得我们只需查看括号内的内容怎么办？
$$
\textcolor{lightgray}{\text{数} = 
7 \Biggr( 
    10^{n-2}k + 
    10^{n-3}m +
    \cdots +
    10c + 
    b
\Biggr) +
3 \biggr( 
    10^{n-2}k + 10^{n-3}m + \cdots + 10c + b
\biggr) +} \space
a
$$

为了不遗漏任何部分，

我们需要 `a` 可以表达为两部分，其中 

- 一部分必须 `整合到` 含有 `7的倍数` 的表达式中
- 而另一部分 `整合到` 含有 `3的倍数` 的表达式中

像这样
$$
\textcolor{lightgray}{\text{数} = 
7 \Biggr( 
    10^{n-2}k + 
    10^{n-3}m +
    \cdots +
    10c + 
    b
\Biggr) +
3 \biggr( 
    10^{n-2}k + 10^{n-3}m + \cdots + 10c + b
\biggr) +} \space
(7a - 6a)
$$

整合到表达式中
$$
\textcolor{lightgray}{\text{数} = 
7 \Biggr( 
    10^{n-2}k + 
    10^{n-3}m +
    \cdots +
    10c + 
    b} +
    a
\textcolor{lightgray}{\Biggr) +
3 \biggr( 
    10^{n-2}k + 10^{n-3}m + \cdots + 10c + b} - 2a
\textcolor{lightgray}{\biggr)}
$$

现在我们只需查看括号中的表达式即可确定 `可否被7整除`
$$
\textcolor{lightgray}{\text{数} = 
7 \Biggr( 
    10^{n-2}k + 
    10^{n-3}m +
    \cdots +
    10c + 
    b +
    a
\Biggr) +
3 \biggr(} 
    \underline{10^{n-2}k + 
    10^{n-3}m + 
    \cdots + 
    10c + 
    b} - 
    2a
\textcolor{lightgray}{\biggr)}
$$

意识到它包含 `数本身的所有数字` 直到 `第 n-1 位` 减去 `2a`
$$
\boxed{
    \underbrace{\overline{k \space m \space \cdots \space c \space b \space}}_{\text{n-1 位}} - 2a
}
$$


# 可除性规则3

因此，判断`整个数字能否被7整除`的条件是：

$$
\boxed{
    \phantom{\biggr(} 
        \underline{
            \textbf{去掉个位数后剩余的数字}
        } 
        \space 
        \textbf{减去} 
        \space 
        \underline{\textbf{个位数字的两倍}} 
        \space 
        \textbf{，所得结果必须能被7整除}
    \phantom{\biggr)}
}
$$
