# 数的构成

$$
\text{数} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n位}} =
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

我们来看看“10的幂”

# 10的幂次分解  

从$10^0$开始  
- 它等于`1`  

$$
10^{0} = 1  
$$  

我们无法`分割`或`因式分解`它，所以继续看$10^1$  

## 有意的分割  

注意到`10`可以`分割`为`包含4的商`和`余数2`  

$$
\boxed{  
    \phantom{\big(}  
        10 = 4 \cdot 2 + 2  
    \phantom{\big)}  
}  
$$  

对于`下一个10的幂次`，我们乘以`10`得到  

$$
100 = 4 \cdot 2 \cdot 10 + 2 \cdot \boxed{10}  
$$  

注意到分割中的`10`，可以像之前一样拆分  

$$
\begin{align}  
    100 &= 4 \cdot 2 \cdot 10 + \underline{2 \space \big( 4 \cdot 2 + 2 \big)} \\  
    &= 4 \cdot 2 \cdot 10 + 4 \cdot 2^2 + 2^2 \\  
\end{align}  
$$  

因此，  

$$
\boxed{  
    \phantom{\big(}  
        100 = 4 \cdot 2 \cdot 10 + 4 \cdot 2^2 + 2^2  
    \phantom{\big)}  
}  
$$  

这个规律在更高次幂也成立，  

$$
1000 = 4 \cdot 2 \cdot 10^2 + 4 \cdot 2^2 \cdot 10 + 2^2 \cdot \boxed{10}  
$$  

可以像之前一样简化  

$$
\begin{align}  
    1000 &= 4 \cdot 2 \cdot 10^2 + 4 \cdot 2^2 \cdot 10 + 2^2 \cdot \big( 4 \cdot 2 + 2 \big) \\  
    &= 4 \cdot 2 \cdot 10^2 + 4 \cdot 2^2 \cdot 10 + 4 \cdot 2^3 + 2^3  
\end{align}  
$$  

$$
\boxed{  
    \phantom{\big(}  
        1000 = 4 \cdot 2 \cdot 10^2 + 4 \cdot 2^2 \cdot 10 + 4 \cdot 2^3 + 2^3  
    \phantom{\big)}  
}  
$$  

## 分割模式的推广  

因此，我们得到这个从`2次幂`开始到`n次幂`都成立的模式  

$$
\begin{align}  
    10 &= 4 \cdot 2 + 2 \\  
    100 &= 4 \cdot 2 \cdot 10 + 4 \cdot 2^2 + 2^2 \\  
    1000 &= 4 \cdot 2 \cdot 10^2 + 4 \cdot 2^2 \cdot 10 + 4 \cdot 2^3 + 2^3 \\  
    &\vdots \\  
    10^n &= 4 \cdot 2^1 \cdot 10^{n-1} + 4 \cdot 2^2 \cdot 10^{n-2} + \cdots + 4 \cdot 2^{n} \cdot 10^{0} + 2^n \\  
\end{align}  
$$  

因此，整体表达式为  

$$
\boxed{  
    \phantom{\Big(}  
        10^n = 4 \biggr( 2^1 \cdot 10^{n-1} + 2^2 \cdot 10^{n-2} + \cdots + 2^{n} \cdot 10^{0} \biggr) + 2^n  
    \phantom{\Big)}  
}  
$$  

# 覆盖大多数情况的特殊条件  

注意到我们有$2^n$  

$$
\textcolor{lightgray}{  
    4  
    \biggr( 2^1 \cdot 10^{n-1} + 2^2 \cdot 10^{n-2} + \cdots + 2^{n} \cdot 10^{0} \biggr)  
} + 2^{n}  
$$  

而`4`也是$2^2$  

$$
2^2  
\textcolor{lightgray}{  
    \biggr( 2^1 \cdot 10^{n-1} + 2^2 \cdot 10^{n-2} + \cdots + 2^{n} \cdot 10^{0} \biggr)  
} + 2^{n}  
$$  

因此，当$n \ge 2$时  

$$
2^2 \space \biggr( 2^1 \cdot 10^{n-1} + 2^2 \cdot 10^{n-2} + \cdots + 2^{n} \cdot 10^{0} \biggr) + 2^{2} \cdot 2^{n-2}  
$$  

我们可以将`2的幂次表达式`合并到`4的倍数`中  
> 使其成为`4的倍数`的`单一表达式`  

$$
2^2 \space \biggr( 2^1 \cdot 10^{n-1} + 2^2 \cdot 10^{n-2} + \cdots + 2^{n} \cdot 10^{0} + 2^{n-2} \biggr)  
$$  

因此，对于$n \ge 2$的一般表达式为  

$$
\underline{  
    \phantom{\Biggr(}  
        10^{n} = 4 \biggr( 2^1 \cdot 10^{n-1} + 2^2 \cdot 10^{n-2} + \cdots + 2^{n} \cdot 10^{0} + 2^{n-2} \biggr)  
    \phantom{\Biggr)}  
}  
$$  

# 通用表达式的应用  

这个表达式基本上适用于所有`10的幂次`  

> 当$n \ge 2$时  
>  
> $$
\boxed{  
    \phantom{\Biggr(}  
        10^{n} = 4 \biggr( 2^1 \cdot 10^{n-1} + 2^2 \cdot 10^{n-2} + \cdots + 2^{n} \cdot 10^{0} + 2^{n-2} \biggr)  
    \phantom{\Biggr)}  
}  
$$  
>  
> 当$n = 1$时  
>  
> $$
\boxed{  
    \phantom{\big(}  
    10^1 = 4 \cdot 2 + 2  
    \phantom{\big)}  
}  
$$  

# 按照结构来看

我们再回过头来看看这个数的结构：

$$
\text{数} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n位}} = 
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

## 按位可除性分析

因为我们知道从第二幂开始的所有 `10的幂` 都是 `4的倍数`

$$
\text{数} = \underbrace{\overline{k \space m \space \cdots \space c \space \textcolor{lightgray}{b \space a}}}_{\text{n位}} = 
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} \textcolor{lightgray}{+ 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}}
$$

所以我们只需要检查最后两位（即十位和个位组成的数）是否能被 `4` 整除即可。

$$
\text{数} = \underbrace{\overline{\textcolor{lightgray}{k \space m \space \cdots \space c} \space b \space a}}_{\text{n位}} = 
\textcolor{lightgray}{\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

# 可除性规则 1

因此，要判断整个数是否能被 `4` 整除，  


$$
\boxed{
    \phantom{\biggr(} 
        \textbf{最后两位组成的数必须是4的倍数}
    \phantom{\biggr)}
}
$$

# 数字的因式分解

让我们再看一下数字的构成

$$
\text{数字} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n位数}} =
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

我们知道

> 对于$n \ge 2$
>
> $$
10^{n} = 4 \biggr( 2^1 \cdot 10^{n-1} + 2^2 \cdot 10^{n-2} + \cdots + 2^{n} \cdot 10^{0} + 2^{n-2} \biggr)
$$
> 
> 对于$n = 1$
>
> $$
10^1 = 4 \cdot 2 + 2
$$

## 简化大表达式

实际上，既然我们知道这个表达式是`8的倍数`

$$
\textcolor{lightgray}{10^n =} 4 \biggr( 2^1 \cdot 10^{n-1} + 2^2 \cdot 10^{n-2} + \cdots + 2^{n} \cdot 10^{0} + 2^{n-2} \biggr)
$$

我们可以这样简化括号内的表达式

- 其中$q_{n} = 2^1 \cdot 10^{n-1} + 2^2 \cdot 10^{n-2} + \cdots + 2^{n} \cdot 10^{0} + 2^{n-2}$

>
> 对于$n \ge 2$
>
> $$
\boxed{
    \phantom{\big(}
        10^n = 4q_{n}
    \phantom{\big)}
}
$$

## 代数运算

我们将其因式分解

$$
\text{数字} = 
k \space \biggr( 4q_{n-1} \biggr) + 
m \space \biggr( 4q_{n-2} \biggr) + 
\cdots + 
c \space \biggr( 4q_{2} \biggr) + 
b \space \biggr( 4 \cdot 2 + 2 \biggr) + 
a
$$

简化后得到

$$
\text{数字} = 
\boxed{ \phantom{\biggr(} 4k \cdot q_{n-1} \phantom{\biggr)}} +
\boxed{ \phantom{\biggr(} 4m \cdot q_{n-2} \phantom{\biggr)}} + 
\cdots +
\boxed{\phantom{\biggr(} 4c \cdot q_{2} \phantom{\biggr)}}  + 
\boxed{\phantom{\biggr(} 4 \cdot 2b + 2b  \phantom{\biggr)}} + 
\boxed{\phantom{\biggr(} a \phantom{\biggr)}}
$$

重新分组如下

$$
\text{数字} = 
\Biggr( 
    4 \cdot kq_{n-1} + 
    4 \cdot mq_{n-2} +
    \cdots +
    4 \cdot cq_{2} +
    4 \cdot 2b
\Biggr) +
\biggr( 
    2b +
    a
\biggr)
$$

按`4`重新分组得到

$$
\boxed{
    \text{数字} = 
    4 \Biggr( 
        kq_{n-1} + 
        mq_{n-2} +
        \cdots +
        cq_{2} +
        2b
    \Biggr) +
    \biggr( 
        2b +
        a
    \biggr)
}
$$

# 余数分析

因此，我们看到数字的`商`是`4的倍数`
- 因此它可以`被4整除`

$$
\textcolor{lightgray}{\text{数字} =} 
    4 \biggr( 
    kq_{n-3} + 
    mq_{n-4} +
    \cdots +
    cq_{0} + 
    2b
    \biggr) 
    \textcolor{lightgray}{+
    \biggr(
    2b +
    a
    \biggr)
    }
$$

因此，要使`整个数字能被4整除`，`余数`也必须是`4的倍数`

$$
\textcolor{lightgray}{\text{数字} = 
    4 \biggr( 
    kq_{n-3} + 
    mq_{n-4} +
    \cdots +
    cq_{0} + 
    2b
    \biggr)} +
    \biggr(
    2b +
    a
    \biggr)
$$

# 可除性规则 2

因此，要判断整个数是否能被 `4` 整除，  


$$
\boxed{
    \phantom{\biggr(} 
        \textbf{该数: } \underline{\textbf{倒数第二位的两倍}} \textbf{ 加上 } \underline{\textbf{最后一位数字}} \textbf{，必须能被4整除}
    \phantom{\biggr)}
}
$$
