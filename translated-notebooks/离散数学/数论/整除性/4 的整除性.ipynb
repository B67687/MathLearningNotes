# 数的构成

$$
\text{数} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n位}} =
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

我们来看看“10的幂”

# 对10的幂进行因式分解

从 $10^0$ 开始  
- 它等于 `1`

$$
10^{0} = 1
$$

这个数既不能“分割”也不能“因式分解”，所以我们继续看下一个，即 $10^1$

## 有意识的拆分

注意到 `10` 可以被“拆分为”一个包含 `4` 的商和余数为 `2` 的形式：

$$
\boxed{ \phantom{\big(} 10 = \underline{4 \cdot 2 + 2} \phantom{\big)} }
$$

如果我们继续到下一个10的幂，我们将前面的结果乘以 `10` 得到：

$$
100 = 4 \cdot 20 + 2 \cdot \boxed{10}
$$

注意到了其中的 `10`，我们可以像之前那样对它进一步拆分：

$$
100 =  4 \cdot 20 + \underline{2 \space \big( 4 \cdot 2 + 2 \big)}
$$

然后按 `4` 进行重组：

$$
\begin{align}
    100 &= \textcolor{lightgray}{4 \cdot 20}  + 4 \space (2 \cdot 2) + 2 \cdot 2 \\  
    &= \textcolor{lightgray}{4 \cdot 20} + 4 \cdot 4 + 4 \cdot 1 \\
    &= 4 \space (20 + 4 + 1) \\
    &= 4 \cdot 25
\end{align}
$$

因此，

$$
\boxed{
    \phantom{\big(} 100 = \underline{4 \cdot 25} \phantom{\big)}
}
$$

这种模式在接下来的幂次中依然成立：

$$
\boxed{ \phantom{\big(} 1000 = \underline{4 \cdot 250} \phantom{\big(}}
$$

再下一项：

$$
\boxed{ \phantom{\big(} 10000 = \underline{4 \cdot 2500} \phantom{\big(}}
$$

## 这种拆分方式的一般化表达

于是我们得到了如下从第二幂开始一直到某个 `n` 的规律性模式：

$$
\begin{align}
    100 &= 4 \cdot 25 \\
    1000 &= 4 \cdot 250 \\
    10000 &= 4 \cdot 2500 \\
    &\vdots \\
    10^n &= 4 \space \big( 25 \cdot 10^{n-2} \big)
\end{align}
$$

# 按照结构来看

我们再回过头来看看这个数的结构：

$$
\text{数} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n位}} = 
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

## 按位可除性分析

因为我们知道从第二幂开始的所有 `10的幂` 都是 `4的倍数`

$$
\text{数} = \underbrace{\overline{k \space m \space \cdots \space c \space \textcolor{lightgray}{b \space a}}}_{\text{n位}} = 
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} \textcolor{lightgray}{+ 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}}
$$

所以我们只需要检查最后两位（即十位和个位组成的数）是否能被 `4` 整除即可。

$$
\text{数} = \underbrace{\overline{\textcolor{lightgray}{k \space m \space \cdots \space c} \space b \space a}}_{\text{n位}} = 
\textcolor{lightgray}{\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

# 可除性规则 1

因此，要判断整个数是否能被 `4` 整除，  


$$
\boxed{
    \phantom{\biggr(} 
        \textbf{最后两位组成的数必须是4的倍数}
    \phantom{\biggr)}
}
$$

# 对数进行因式分解

我们再次看看这个数：

$$
\text{数} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n位}} = 
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

我们知道以下关系成立：
- 其中 `n ≥ 2`

$$
10^{n} = 4 \space \big( 25 \cdot 10^{n-2} \big)
$$

我们将它代入原式中：

$$
\text{数} = 
k \biggr( 4 \space \big( 25 \cdot 10^{n-3} \big) \biggr) + 
m \biggr( 4 \space \big( 25 \cdot 10^{n-4} \big) \biggr) + 
\cdots + 
c \Biggr( 4 \space \big( 25 \cdot 10^0 \big)  \Biggr) + 
b \Biggr( 4 \cdot 2 + 2 \Biggr) + 
a
$$

将所有的 `25 × 10的幂` 表示为：

$$
q_{n} = 25 \cdot 10^{n} 
$$

于是我们得到：

$$
\text{数} = 
k \biggr( 4 \space q_{n-3} \biggr) + 
m \biggr( 4 \space q_{n-4} \biggr) + 
\cdots + 
c \Biggr( 4 \space q_{0}  \Biggr) + 
b \Biggr( 4 \cdot 2 + 2 \Biggr) + 
a
$$

将 `b` 的部分拆成含有 `4` 的项与不含 `4` 的项：

$$
\textcolor{lightgray}{\text{数} = 
k \biggr( 4 \space q_{n-3} \biggr) + 
m \biggr( 4 \space q_{n-4} \biggr) + 
\cdots + 
c \Biggr( 4 \space q_{0} \Biggr)} + 
b \Biggr( 4 \cdot 2 \Biggr) + 
2b \textcolor{lightgray}{+ 
a}
$$

提取出公共因子 `4` 后得到：

$$
\boxed{
    \text{数} = 
    4 \biggr( 
    kq_{n-3} + 
    mq_{n-4} +
    \cdots +
    cq_{0} + 
    2b
    \biggr) +
    2b +
    a
}
$$

# 余数分析

因此，我们看到这个数的“商”是一个 `4` 的倍数  
- 所以它是可以被 `4` 整除的部分

$$
\textcolor{lightgray}{\text{数} =} 
    4 \biggr( 
    kq_{n-3} + 
    mq_{n-4} +
    \cdots +
    cq_{0} + 
    2b
    \biggr) \textcolor{lightgray}{+
    2b +
    a}
$$

所以为了让整个数能被 `4` 整除，它的“余数”也必须是 `4` 的倍数：

$$
\textcolor{lightgray}{\text{数} = 
    4 \biggr( 
    kq_{n-3} + 
    mq_{n-4} +
    \cdots +
    cq_{0} + 
    2b
    \biggr)} +
    2b +
    a
$$

# 可除性规则 2

因此，要判断整个数是否能被 `4` 整除，  


$$
\boxed{
    \phantom{\biggr(} 
        \textbf{该数: } \underline{\textbf{倒数第二位的两倍}} \textbf{ 加上 } \underline{\textbf{最后一位数字}} \textbf{，必须能被4整除}
    \phantom{\biggr)}
}
$$
