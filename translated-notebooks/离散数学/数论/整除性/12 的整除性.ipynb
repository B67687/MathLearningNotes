# 合数

`12` 是一个 `合数`，它由 `4` 和 `3` 组成

$$
12 = 4 \cdot 3
$$

这意味着只要一个数 `能被4和3整除`，它就 `能被12整除`

$$
\underline{\text{能被12整除}} = \boxed{\text{能被3整除}} \ \& \ \boxed{\text{能被4整除}} 
$$

我们已经在之前的 `jupyter notebook` 中 `推导出以下条件`：
- `能被3整除` 的条件  
- `能被4整除` 的条件  

因此，我们直接将它们搬到这里

## 能被3整除

要使一个数 `能被3整除`，必须满足以下条件：

$$
\phantom{\biggr(} 
    \textbf{所有数字之和是3的倍数}
\phantom{\biggr)}
$$

## 能被4整除

要使一个数 `能被4整除`，必须满足以下任意一个条件：

$$
\phantom{\biggr(} 
    \textbf{十位及之前的数必须是4的倍数}
\phantom{\biggr)}
$$

$$
\textbf{或}
$$

$$
\phantom{\biggr(} 
    \underline{\textbf{倒数第二位的两倍}} \textbf{ 加上 } \underline{\textbf{最后一位}} \textbf{ 的和必须能被4整除}
\phantom{\biggr)}
$$

# 结论

因此，要使一个数 `能被12整除`，必须满足以下条件：

---

- **能被3整除**

$$
\boxed{
    \phantom{\biggr(} 
        \textbf{所有数字之和是3的倍数}
    \phantom{\biggr)}
}
$$

- **能被4整除**

$$
\boxed{
    \phantom{\biggr(} 
        \textbf{十位及之前的数必须是4的倍数}
    \phantom{\biggr)}
}
$$

$$
\textbf{或}
$$

$$
\boxed{
    \phantom{\biggr(} 
        \textbf{： } \underline{\textbf{倒数第二位的两倍}} \textbf{ 加上 } \underline{\textbf{最后一位}} \textbf{ 的和必须能被4整除}
    \phantom{\biggr)}
}
$$

> ✅ 只有当 **这些条件都满足** 时，这个数才 **能被12整除**
