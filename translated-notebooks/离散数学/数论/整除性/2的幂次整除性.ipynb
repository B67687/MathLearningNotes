# 数的构成

$$
\text{数字} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n 位数}} =
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

让我们观察`10的幂次`

# 10的幂次分解

从$10^0$开始
- 它等于`1`

$$
10^{0} = 1
$$

我们无法`分解`或`因式分解`它，所以继续看$10^1$

## 有意的分解

注意到`10`可以`分解`为`2`和`5`

$$
\boxed{ 
    \phantom{\big(} 
    10 = 2 \cdot 5
    \phantom{\big)}
}
$$

如果我们继续看`下一个10的幂次`，乘以`10`得到

$$
100 = \biggr( 2 \cdot 5 \biggr) \cdot \boxed{10}
$$

注意到分解中的`10`，可以像之前一样拆分

$$
100 = \biggr( 2 \cdot 5 \biggr) \cdot \biggr( 2 \cdot 5 \biggr)
$$

于是得到

$$
\boxed{
    \phantom{\big(} 
    100 = 2^2 \cdot 5^2
    \phantom{\big)}
}
$$

这个规律在更高幂次也成立，

$$
1000 = \biggr( 2^2 \cdot 5^2 \biggr) \cdot \boxed{10}
$$

拆分它
$$
1000 = \biggr( 2^2 \cdot 5^2 \biggr) \cdot \biggr( 2 \cdot 5 \biggr)
$$

于是得到
$$
\boxed{ 
    \phantom{\big(} 
    1000 = 2^3 \cdot 5^3
    \phantom{\big(}
}
$$

## 分解的推广

因此，我们得到这个从`2次幂`开始到`n次幂`都成立的模式

$$
\begin{align}
    10 &= 2 \cdot 5 \\
    100 &= 2^2 \cdot 5^2 \\
    1000 &= 2^3 \cdot 5^3 \\
    &\vdots \\
    10^n &= 2^n \cdot 5^n \\
\end{align}
$$

因此，整体表达式为

$$
\boxed{
    \phantom{\Big(}
    10^n = 2^n \cdot 5^n
    \phantom{\Big)}
}
$$


# 回归数字结构

$$

\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

## 按位整除

由于我们知道从`t次幂`开始的每一个`10的幂次`都能被$2^t$`整除`

$$
\text{数字} = \underbrace{\overline{k \space m \space \cdots \space h \space \textcolor{lightgray}{\cdots \space c \space b \space a}}}_{\text{n 位数}} = 
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{h \cdot 10^{t}} \textcolor{lightgray}{+
\cdots +
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}}
$$

我们只需要检查`前(t-1)位`是否能被$2^t$`整除`

$$
\text{数字} = \underbrace{\overline{\textcolor{lightgray}{k \space m \space \cdots \space h} \space \cdots \space c \space b \space a}}_{\text{n 位数}} = 
\textcolor{lightgray}{\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{h \cdot 10^{t}}} +
\cdots +
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

# 2的幂次整除规则

$$
\boxed{
    \phantom{\biggr(}
    \textbf{前(t-1)位数字能被} \ 2^t \ \textbf{整除}
    \phantom{\biggr)}
}
$$
