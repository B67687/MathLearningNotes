# 数字的构成

$$
\text{数字} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n 位}} =
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

我们来看看 `10 的幂`


# 分解 10 的幂

从“被3整除”的笔记本中我们知道：

- 所有非负整数次幂的 `10` 都可以这样拆分：

$$
10^{n} = \underline{
    3^2 \cdot 10^{n-1} + 3^2 \cdot 10^{n-2} + \cdots + 3^2 \cdot 10^{1} + 3^2 \cdot 10^{0} + 1
}
$$

当我们比“被3整除”的方式更进一步地进行简化时，我们得到：

$$
\boxed{
    10^{n} = 9 \cdot \underbrace{\overline{11\cdots11}}_{\text{n 位}} + 1
}
$$

我们发现 `10 的幂` 可以被拆分成：
- 一个能被 `9` 整除的部分

$$
\textcolor{lightgray}{10^{n} =} 9 \cdot \underbrace{\overline{11\cdots11}}_{\text{n 位}} \textcolor{lightgray}{+ 1}
$$

- 然后是一个 `1`

$$
\textcolor{lightgray}{10^{n} = 9 \cdot \underbrace{\overline{11\cdots11}}_{\text{n 位}}} + 1
$$


# 对任意数字的因式分解

使用与“被3整除”笔记相同的 `推导逻辑`，我们会得到：

> 其中 $q_{n} = \underbrace{\overline{11\cdots11}}_{\text{n 位}}$

$$
\text{数字} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n 位}} =
9 \biggr( 
k \cdot q_{n-1} + 
m \cdot q_{n-2} +
\cdots +
c \cdot q_{2} + 
b\cdot q_{1}
\biggr) +
\big( 
k + m + \cdots + c + b + a
\big)
$$

# 结论

因此，我们看到这个部分是 `9 的倍数`
- 因此它是 `可被 9 整除的`

$$
\textcolor{lightgray}{\text{数字} = }
9 \biggr( 
    kq_{n-1} + 
    mq_{n-2} +
    \cdots +
    cq_{2} + 
    bq_{1}
\biggr)
\textcolor{lightgray}{+
\big( k + m + \cdots + c + b + a \big)}
$$

因此，要使整个数字 `能被 9 整除`，第二部分也必须是 `9 的倍数`

$$
\textcolor{lightgray}{\text{数字} = 
9 \biggr( 
    kq_{n-1} + 
    mq_{n-2} +
    \cdots +
    cq_{2} + 
    bq_{1}
\biggr)} +
\big( k + m + \cdots + c + b + a \big)
$$

我们注意到这部分包含了数字本身的所有位数：

$$
\text{数字} \space = \space \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n 位}}
$$

---

因此，判断一个 `数字是否能被 9 整除`，只需要：

$$
\text{所有位数之和是 9 的倍数}
$$

# 递归迭代

由于一个数字的各位之和总是 `小于或等于数字本身`。

---

我们可以先计算原始数字的各位之和，看看是否明显能被 `9` 整除。
- 如果不明显，我们可以继续计算这个和的各位之和，并判断它是否能被 `9` 整除。

> 我们可以不断重复这个操作，直到某一次迭代的结果明显能被 `9` 整除！

# 示例

让我们用一个例子来验证！

判断这个数字是否能被 `9` 整除：

$$
535137714
$$

我们计算它的各位之和：

$$
\begin{align}
    5 + 3 + 5 + 1 + 3 + 7 + 7 + 1 + 4 &= (5 + 5) + (3 + 7) + (3 + 7) + (1 + 1 + 4) \\
    &= 10 + 10 + 10 + 6 \\
    &= 36
\end{align}
$$

我们知道：

$$
36 = 9 \cdot 4 
$$

如果这还不够明显，我们可以再做一次！
- 第二次迭代：

$$
3 + 6 = 9
$$

我们知道 `9` 肯定是 `9 的倍数`

$$
9 = 9 \cdot 1
$$

这是所有能被 `9` 整除的数字在最后一步都会得到的结果！

---

因此，这个数字是 `能被 9 整除的`

$$
535137714 = 9 \cdot 59459746
$$

# 合数

`9` 是一个由两个 `3` 构成的合数：

$$
9 = 3^2
$$

这意味着只要一个数字 `能被 3 整除两次`，它就能被 `9` 整除：

$$
\underline{\text{能被 9 整除}} = \boxed{\text{能被 3 整除两次}} 
$$

我们已经在之前的 `jupyter notebook` 中推导出了 `被3整除` 的条件。

现在我们将这些内容移过来：

## 被3整除

要使一个数字 `能被 3 整除`，必须满足以下条件：

$$
\text{所有位数之和是 3 的倍数}
$$

## 被9整除

因此，要使一个数字 `能被 9 整除`

$$
\text{在两次迭代下，所有位数之和是 3 的倍数}
$$

换句话说：
> 如果 **所有位数之和** 能被 3 整除，而且 **该和的各位之和** 也能被 3 整除，则这个数字能被 9 整除。
