# 一个数的构成

$$
\text{数} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n位}} =
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

让我们看一下 `10 的幂`

# 分解 10 的幂

我们从 $10^0$ 开始  
- 它等于 `1`

$$
10^{0} = 1
$$

我们无法对它进行 `分割` 或 `因式分解`，因此我们继续看 $10^1$

## 有意分割

注意 `10` 可以以 `11` 和 `1` 表达

$$
\boxed{ 
    \phantom{\big(} 
    10 = \underline{11 - 1} 
    \phantom{\big)} 
}
$$

如果我们继续到下一个 `10 的幂`，我们将乘以 `10` 得到

$$
\boxed{ 
    \phantom{\big(} 
    100 = \underline{11 \cdot 10 - 10} 
    \phantom{\big)} 
}
$$

对于下一个 `10 的幂` 也是一样

$$
\boxed{ 
    \phantom{\big(} 
    1000 = \underline{11 \cdot 10^2 - 10^2} 
    \phantom{\big)} 
}
$$

## 此类分割的一般化

因此，我们得到了一个从 `第一个幂开始` 到某个 `n` 都重复的模式

$$
\begin{align}
    10 &= 11 - 1 \\ 
    100 &= 11 \cdot 10 - 10 \\
    1000 &= 11 \cdot 10^2 - 10^2 \\
    &\vdots \\
    10^n &= 11 \cdot 10^{n-1} - 10^{n-1} \\
\end{align}
$$

# 因式分解这个数

再来看一遍这个数

$$
\text{数} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n位}} = 
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

因为我们知道这个  
- 其中 `n` 从 `1` 开始增加

$$
10^n = 11 \cdot 10^{n-1} - 10^{n-1}
$$

我们将它代入因式分解中

$$
\text{数} = 
\space k \space \biggr( 11 \cdot 10^{n-2} - 10^{n-2} \biggr) + 
m \space \biggr( 11 \cdot 10^{n-3} - 10^{n-3} \biggr) + 
\cdots + 
c \space \biggr( 11 \cdot 10 - 10 \biggr) + 
b \space \biggr( 11 - 1 \biggr) + 
a
$$

简化后，我们得到

$$
\text{数} = 
\boxed{ \phantom{\biggr(} 11 \cdot 10^{n-2}k - 10^{n-2}k \phantom{\biggr)}} +
\boxed{ \phantom{\biggr(} 11 \cdot 10^{n-3}m - 10^{n-3}m \phantom{\biggr)}} + 
\cdots +
\boxed{\phantom{\biggr(} 11 \cdot 10c - 10c \phantom{\biggr)}}  + 
\boxed{\phantom{\biggr(} 11b - b  \phantom{\biggr)}} + 
\boxed{\phantom{\biggr(} a \phantom{\biggr)}}
$$

像这样重新分组

$$
\text{数} = 
\Biggr( 
    11 \cdot 10^{n-2}k + 
    11 \cdot 10^{n-3}m +
    \cdots +
    11 \cdot 10c + 
    11b
\Biggr) -
\biggr( 
    10^{n-2}k + 10^{n-3}m + \cdots + 10c + b
\biggr) +
    a
$$

将这些组按 `11` 重新分组，我们得到

$$
\text{数} = 
11 \Biggr( 
    10^{n-2}k + 
    10^{n-3}m +
    \cdots +
    10c + 
    b
\Biggr) -
\biggr( 
    10^{n-2}k + 10^{n-3}m + \cdots + 10c + b
\biggr) +
a
$$

注意到这也是该数的一部分（在 `十进制` 中）

$$
\text{数} = 
11 \cdot \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} \space -
\underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} +
a
$$


# 余数分析

因此，我们看到数的这一部分是 `11 的倍数`  
- 从而使其 `可被11整除`

$$
\textcolor{lightgray}{\text{数} = }
    11 \cdot \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}}
\space
\textcolor{lightgray}{-}
\space
\textcolor{lightgray}{
    \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} + a
}
$$

因此，要使 `整个数能被11整除`，这一部分也必须是 `11 的倍数`

$$
\textcolor{lightgray}{\text{数} =
    11 \cdot \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}}} \space -
    \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} + a
$$

但它仍然是 `两个表达式`，我们将其简化为 `一个表达式`

$$
\textcolor{lightgray}{\text{数} =
    11 \cdot \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}}} \space -
    \biggr( \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} - a \biggr)
$$

这样我们只需查看 `绝对表达式` 即可

$$
\textcolor{lightgray}{\text{数} =
    11 \cdot \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} \space -}
    \biggr( \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} - a \biggr)
$$

就是这个

$$
\boxed{
    \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} - a
}
$$

# 整除性规则1

因此，要确定 `整个数能否被11整除`,

$$
\boxed{
    \phantom{\biggr(} 
        \textbf{不包括个位的数减去个位上的数，必须能被11整除}
    \phantom{\biggr)}
}
$$

# 强化分割

注意到 `10` 可以用 `11` 和 `1` 来表示

$$
\boxed{ 
    \phantom{\big(} 
    10 = \underline{11 - 1} 
    \phantom{\big)} 
}
$$

如果我们继续到下一个 `10 的幂`，我们将乘以 `10` 得到

$$ 
100 = 11 \cdot 10 - \boxed{10}
$$

注意这个单独的表达式 `10` 是唯一一个不是以 `11` 表达的部分  
- 我们可以像之前一样对它进行 `拆分`

$$ 
\begin{align}
    100 &= 11 \cdot 10 - (11 - 1) \\
    &= 11 \cdot 10 - 11 + 1
\end{align}
$$

从而得到

$$
\boxed{ 
    \phantom{\big(} 
    100 = \underline{11 \cdot 10 - 11 + 1} 
    \phantom{\big)} 
}
$$

继续到下一个 `10 的幂`，我们有

$$
1000 = 11 \cdot 10^2 - 11 \cdot 10 + 1 \cdot \boxed{10}
$$

同样地，表达式 `10` 是唯一一个不是以 `11` 表达的部分  
- 我们对它进行 `拆分`

$$
\begin{align}
    1000 &= 11 \cdot 10^2 - 11 \cdot 10 + (11 - 1) \\
    &= 11 \cdot 10^2 - 11 \cdot 10 + 11 - 1
\end{align}
$$

然后我们得到

$$
\boxed{ 
    \phantom{\big(} 
    1000 = \underline{11 \cdot 10^2 - 11 \cdot 10 + 11 - 1} 
    \phantom{\big)} 
}
$$

## 此类分割的一般化

因此，我们得到了一个从 `第一个幂开始` 到某个 `n` 都重复的模式

$$
\begin{align}
    10 &= 11 - 1 \\ 
    100 &= 11 \cdot 10 - 11 + 1 \\
    1000 &= 11 \cdot 10^2 - 11 \cdot 10 + 11 - 1 \\
    &\vdots \\
        10^n &= 11 \cdot 10^{n-1} - 11 \cdot 10^{n-2} + \cdots \pm 11 \cdot 10^{0} \space \mp 1 \\
\end{align}
$$

我们可以进一步将一般表达式简化为：

- 对于 **奇数 n**

$$
\boxed{
    10^n = 11 \biggr( 10^{n-1} - 10^{n-2} + \cdots - 1 \biggr) + 1
}
$$

- 对于 **偶数 n**

$$
\boxed{
    10^n = 11 \biggr( 10^{n-1} - 10^{n-2} + \cdots + 1 \biggr) - 1
}
$$

- 总体形式

$$
\boxed{
    10^n = 11 \biggr( 10^{n-1} - 10^{n-2} + \cdots \pm 1 \biggr) \mp 1
}
$$


# 因式分解这个数

再来看一遍这个数

$$
\text{数} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n位}} = 
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

因为我们知道这个  
- 其中 `n` 从 `1` 开始增加

$$
10^n = 11 \biggr( 10^{n-1} - 10^{n-2} + \cdots \pm 1 \biggr) \mp 1
$$

---

实际上，由于我们知道这部分是 `11 的倍数`

$$
\textcolor{lightgray}{10^n =} 11 \biggr( 10^{n-1} - 10^{n-2} + \cdots \pm 1 \biggr) \textcolor{lightgray}{\mp} \textcolor{lightgray}{1}
$$

我们可以将括号内的部分简化如下：

> 其中 $q_{n} = 10^{n-1} - 10^{n-2} + \cdots \pm 1$

$$
\boxed{
    10^n = 11q_{n} \mp 1
}
$$

---

我们将它代入因式分解中

$$
\text{数} = 
\space k \space \biggr( 11q_{n-1} \mp 1 \biggr) + 
m \space \biggr( 11q_{n-2} \pm 1 \biggr) + 
\cdots + 
c \space \biggr( 11q_{2} \mp 1 \biggr) + 
b \space \biggr( 11q_{1} \pm 1 \biggr) + 
a
$$

简化后，我们得到

$$
\text{数} = 
\boxed{ \phantom{\biggr(} 11k \cdot q_{n-1} \mp k \phantom{\biggr)}} +
\boxed{ \phantom{\biggr(} 11m \cdot q_{n-2} \pm m \phantom{\biggr)}} + 
\cdots +
\boxed{\phantom{\biggr(} 11c \cdot q_{2} \mp c \phantom{\biggr)}}  + 
\boxed{\phantom{\biggr(} 11b \cdot q_{1} \pm b  \phantom{\biggr)}} + 
\boxed{\phantom{\biggr(} a \phantom{\biggr)}}
$$

像这样重新分组

$$
\text{数} = 
\Biggr( 
    11 \cdot kq_{n-1} + 
    11 \cdot mq_{n-2} +
    \cdots +
    11 \cdot cq_{2} + 
    11 \cdot bq_{1}
\Biggr) +
\biggr( 
    \mp k \pm m \mp \cdots \pm c \mp b
\biggr) + 
a
$$

将这些组按 `11` 重新分组，我们得到

$$
\boxed{
    \text{数} = 
    11 \Biggr( 
        kq_{n-1} + 
        mq_{n-2} +
        \cdots +
        cq_{2} + 
        bq_{1}
    \Biggr) +
    \biggr( 
        \mp k \pm m \mp \cdots \pm c \mp b
    \biggr)
    + a
}
$$

当 `n 是偶数` 时，它是

$$
\boxed{
    \text{数} = 
    11 \Biggr( 
        kq_{n-1} + 
        mq_{n-2} +
        \cdots +
        cq_{2} + 
        bq_{1}
    \Biggr) +
    \biggr( 
        - k + m - \cdots - c + b
    \biggr) + 
    a
}
$$

当 `n 是奇数` 时，它是

$$
\boxed{
    \text{数} = 
    11 \Biggr( 
        kq_{n-1} + 
        mq_{n-2} +
        \cdots +
        cq_{2} + 
        bq_{1}
    \Biggr) +
    \biggr( 
        k - m + \cdots + c - b
    \biggr) +
    a
}
$$

# 余数分析

因此，我们看到数的这一部分是 `11 的倍数`  
- 从而使其 `可被11整除`

$$
\textcolor{lightgray}{\text{数} = }
11 \Biggr( 
        kq_{n-1} + 
        mq_{n-2} +
        \cdots +
        cq_{2} + 
        bq_{1}
    \Biggr)
\textcolor{lightgray}{+
\big( \mp k \pm m \mp \cdots \pm c \mp b \big) + a}
$$

因此，要使 `整个数能被11整除`，这一部分也必须是 `11 的倍数`

$$
\textcolor{lightgray}{\text{数} = 
11 \Biggr( 
        kq_{n-1} + 
        mq_{n-2} +
        \cdots +
        cq_{2} + 
        bq_{1}
    \Biggr)} +
\big( \mp k \pm m \mp \cdots \pm c \mp b \big) + a
$$

注意到它包含了 `数本身的所有数字`，但使用了 `交替操作符`，除了 `a` 外

$$
\text{数} \space = \space \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n位}}
$$

## 偶数和奇数

注意到 `该部分中所有数字组成的表达式能被11整除` 的两种不同情况：

- `n 是偶数`

$$
\underline{- k + m - \cdots + c - b} + a
$$

- `n 是奇数`

$$
\underline{k - m + \cdots + c - b} + a
$$

## 正负相通

请注意，其负值也必须 `能被11整除`

- `n 是偶数`

$$
\underline{k - m + \cdots - c + b} - a
$$

- `n 是奇数`

$$
\underline{- k + m - \cdots - c + b} - a
$$

## 选择正数开头

我们可以选择那些以正数开头的情况，并最终得到：

- `n 是偶数`

$$
\underline{k - m + \cdots - c + b} - a
$$

- `n 是奇数`

$$
\underline{k - m + \cdots + c - b} + a
$$

## 总结

因此，我们可以将其归纳为一条规则：

$$
\boxed{
    k - m + \cdots \pm c \mp b \pm a
}
$$

> 这意味着我们从正的 `k` 开始，然后在加减之间 **交替**

有了这种模式，我们也可以断言，在这个交替和中：
- `奇数位置` 上的数字为正，  
- `偶数位置` 上的数字为负  

因此，我们也可以取 `奇数位置上数字之和` 减去 `偶数位置上数字之和`

# 整除性规则2

因此，要确定 `整个数能否被11整除`，

$$
\boxed{
    \phantom{\biggr(} 
        \textbf{从最大位开始的交替和，必须能被11整除}
    \phantom{\biggr)}
}
$$

或者

$$
\boxed{
    \phantom{\biggr(} 
        \textbf{奇数位置上的数字之和减去偶数位置上的数字之和，必须能被11整除}
    \phantom{\biggr)}
}
$$
