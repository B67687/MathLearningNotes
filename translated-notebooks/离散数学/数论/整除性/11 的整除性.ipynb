# 一个数的构成

$$
\text{数} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n位}} =
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

让我们看一下 `10 的幂`

# 10的幂的因式分解

从$10^0$开始
- 等于`1`

$$
10^{0} = 1
$$

无法对此进行"划分"或"因式分解"，因此继续看$10^1$

## 有意划分

注意到`10`可以表示为`11`和`1`的组合

$$
\boxed{ 
    \phantom{\big(} 
    10 = 11 - 1
    \phantom{\big)} 
}
$$

对于下一个10的幂，乘以`10`得到：

$$
\boxed{ 
    \phantom{\big(} 
    100 = 11 \cdot 10 - 10 
    \phantom{\big)} 
}
$$

同理，对于再下一个10的幂：

$$
\boxed{ 
    \phantom{\big(} 
    1000 = 11 \cdot 10^2 - 10^2 
    \phantom{\big)} 
}
$$

## 此类划分的推广

由此得到从`第1个幂`到`n`的重复模式：

$$
\begin{align}
    10 &= 11 - 1 \\ 
    100 &= 11 \cdot 10 - 10 \\
    1000 &= 11 \cdot 10^2 - 10^2 \\
    &\vdots \\
    10^n &= 11 \cdot 10^{n-1} - 10^{n-1} \\
\end{align}
$$

因此，通用表达式为：

$$
\boxed{
    \phantom{\big(}
    10^n = 11 \cdot 10^{n-1} - 10^{n-1}
    \phantom{\big)}
}
$$

# 因式分解这个数

再来看一遍这个数

$$
\text{数} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n位}} = 
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

因为我们知道这个  
- 其中 `n` 从 `1` 开始增加

$$
10^n = 11 \cdot 10^{n-1} - 10^{n-1}
$$

我们将它代入因式分解中

$$
\text{数} = 
\space k \space \biggr( 11 \cdot 10^{n-2} - 10^{n-2} \biggr) + 
m \space \biggr( 11 \cdot 10^{n-3} - 10^{n-3} \biggr) + 
\cdots + 
c \space \biggr( 11 \cdot 10 - 10 \biggr) + 
b \space \biggr( 11 - 1 \biggr) + 
a
$$

简化后，我们得到

$$
\text{数} = 
\boxed{ \phantom{\biggr(} 11 \cdot 10^{n-2}k - 10^{n-2}k \phantom{\biggr)}} +
\boxed{ \phantom{\biggr(} 11 \cdot 10^{n-3}m - 10^{n-3}m \phantom{\biggr)}} + 
\cdots +
\boxed{\phantom{\biggr(} 11 \cdot 10c - 10c \phantom{\biggr)}}  + 
\boxed{\phantom{\biggr(} 11b - b  \phantom{\biggr)}} + 
\boxed{\phantom{\biggr(} a \phantom{\biggr)}}
$$

像这样重新分组

$$
\text{数} = 
\Biggr( 
    11 \cdot 10^{n-2}k + 
    11 \cdot 10^{n-3}m +
    \cdots +
    11 \cdot 10c + 
    11b
\Biggr) -
\biggr( 
    10^{n-2}k + 10^{n-3}m + \cdots + 10c + b
\biggr) +
    a
$$

将这些组按 `11` 重新分组，我们得到

$$
\text{数} = 
11 \Biggr( 
    10^{n-2}k + 
    10^{n-3}m +
    \cdots +
    10c + 
    b
\Biggr) -
\biggr( 
    10^{n-2}k + 10^{n-3}m + \cdots + 10c + b
\biggr) +
a
$$

注意到这也是该数的一部分（在 `十进制` 中）

$$
\text{数} = 
11 \cdot \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} \space -
\underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} +
a
$$


# 余数分析

因此，我们看到数的这一部分是 `11 的倍数`  
- 从而使其 `可被11整除`

$$
\textcolor{lightgray}{\text{数} = }
    11 \cdot \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}}
\space
\textcolor{lightgray}{-}
\space
\textcolor{lightgray}{
    \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} + a
}
$$

因此，要使 `整个数能被11整除`，这一部分也必须是 `11 的倍数`

$$
\textcolor{lightgray}{\text{数} =
    11 \cdot \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}}} \space -
    \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} + a
$$

但它仍然是 `两个表达式`，我们将其简化为 `一个表达式`

$$
\textcolor{lightgray}{\text{数} =
    11 \cdot \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}}} \space -
    \biggr( \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} - a \biggr)
$$

这样我们只需查看 `绝对表达式` 即可

$$
\textcolor{lightgray}{\text{数} =
    11 \cdot \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} \space -}
    \biggr( \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} - a \biggr)
$$

就是这个

$$
\boxed{
    \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} - a
}
$$

# 整除性规则1

因此，要确定 `整个数能否被11整除`,

$$
\boxed{
    \phantom{\biggr(} 
        \textbf{不包括个位的数减去个位上的数，必须能被11整除}
    \phantom{\biggr)}
}
$$

## 深度划分

注意到`10`可以表示为`11`和`1`的组合

$$
\boxed{ 
    \phantom{\big(} 
        10 = 11 - 1
    \phantom{\big)} 
}
$$

对于下一个10的幂，乘以`10`得到：

$$ 
100 = 11 \cdot 10 - \boxed{10}
$$

注意到`10`是唯一不含`11`的表达式
- 可以像之前那样进行拆分

$$ 
\begin{align}
    100 &= 11 \cdot 10 - (11 - 1) \\
    &= 11 \cdot 10 - 11 + 1
\end{align}
$$

最终得到：

$$
\boxed{ 
    \phantom{\big(} 
        100 = 11 \cdot 10 - 11 + 1
    \phantom{\big)} 
}
$$

对于再下一个10的幂：

$$
1000 = 11 \cdot 10^2 - 11 \cdot 10 + 1 \cdot \boxed{10}
$$

同样地，`10`是唯一不含`11`的表达式
- 进行拆分

$$
\begin{align}
    1000 &= 11 \cdot 10^2 - 11 \cdot 10 + (11 - 1) \\
    &= 11 \cdot 10^2 - 11 \cdot 10 + 11 - 1
\end{align}
$$

得到：

$$
\boxed{ 
    \phantom{\big(} 
        1000 = 11 \cdot 10^2 - 11 \cdot 10 + 11 - 1 
    \phantom{\big)} 
}
$$

## 此类划分的推广

由此得到从`第1个幂`到`n`的通用模式：

$$
\begin{align}
    10 &= 11 - 1 \\ 
    100 &= 11 \cdot 10 - 11 + 1 \\
    1000 &= 11 \cdot 10^2 - 11 \cdot 10 + 11 - 1 \\
    &\vdots \\
        10^n &= 11 \cdot 10^{n-1} - 11 \cdot 10^{n-2} + \cdots \pm 11 \cdot 10^{0} \space \mp 1 \\
\end{align}
$$

可以进一步简化为：

- **通用形式**

$$
\boxed{
    10^n = 11 \biggr( 10^{n-1} - 10^{n-2} + \cdots \pm 1 \biggr) \mp 1
}
$$

---

> 当n为`奇数`时：
>
> $$
\textcolor{lightgray}{
    10^n = 11 \biggr( 10^{n-1} - 10^{n-2} + \cdots
    } 
    - 1
    \textcolor{lightgray}{\biggr)} 
    + 1
$$
>
> 当n为`偶数`时：
>
> $$
\textcolor{lightgray}{
    10^n = 11 \biggr( 10^{n-1} - 10^{n-2} + \cdots
    }
    + 1 
    \textcolor{lightgray}{\biggr)}
    - 1
$$

# 数字的因式分解

重新观察数字表达式：

$$
\text{数字} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n位}} = 
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

已知当n从1开始时的幂表达式：

$$
10^n = 11 \biggr( 10^{n-1} - 10^{n-2} + \cdots \pm 1 \biggr) \mp 1
$$

## 表达式简化

由于该部分已是11的倍数：

$$
\textcolor{lightgray}{10^n =} 11 \biggr( 10^{n-1} - 10^{n-2} + \cdots \pm 1 \biggr) \textcolor{lightgray}{\mp} \textcolor{lightgray}{1}
$$

设$q_{n} = 10^{n-1} - 10^{n-2} + \cdots \pm 1$，则可简化为：

$$
\boxed{
    10^n = 11q_{n} \mp 1
}
$$

## 代数运算

将幂表达式代入数字分解：

$$
\text{数字} = 
\space k \space \biggr( 11q_{n-1} \mp 1 \biggr) + 
m \space \biggr( 11q_{n-2} \pm 1 \biggr) + 
\cdots + 
c \space \biggr( 11q_{2} \mp 1 \biggr) + 
b \space \biggr( 11q_{1} \pm 1 \biggr) + 
a
$$

简化后得到：

$$
\text{数字} = 
\boxed{ \phantom{\biggr(} 11k \cdot q_{n-1} \mp k \phantom{\biggr)}} +
\boxed{ \phantom{\biggr(} 11m \cdot q_{n-2} \pm m \phantom{\biggr)}} + 
\cdots +
\boxed{\phantom{\biggr(} 11c \cdot q_{2} \mp c \phantom{\biggr)}}  + 
\boxed{\phantom{\biggr(} 11b \cdot q_{1} \pm b  \phantom{\biggr)}} + 
\boxed{\phantom{\biggr(} a \phantom{\biggr)}}
$$

重新分组：

$$
\text{数字} = 
\Biggr( 
    11 \cdot kq_{n-1} + 
    11 \cdot mq_{n-2} +
    \cdots +
    11 \cdot cq_{2} + 
    11 \cdot bq_{1}
\Biggr) +
\biggr( 
    \mp k \pm m \mp \cdots \pm c \mp b
\biggr) + 
a
$$

按11合并后：

$$
\text{数字} = 
    11 \Biggr( 
        kq_{n-1} + 
        mq_{n-2} +
        \cdots +
        cq_{2} + 
        bq_{1}
    \Biggr) +
    \biggr( 
        \mp k \pm m \mp \cdots \pm c \mp b
    \biggr)
    + a
$$

---

> 当n为偶数时：
>
> $$
\textcolor{lightgray}{
    \text{数字} = 
    11 \Biggr( 
        kq_{n-1} + 
        mq_{n-2} +
        \cdots +
        cq_{2} + 
        bq_{1}
    \Biggr) +}
    \biggr( 
        - k + m - \cdots - c + b
    \biggr) 
    \textcolor{lightgray}{+a}
$$
> 
> 当n为奇数时：
>
> $$
\textcolor{lightgray}{
    \text{数字} = 
    11 \Biggr( 
        kq_{n-1} + 
        mq_{n-2} +
        \cdots +
        cq_{2} + 
        bq_{1}
    \Biggr) +}
    \biggr( 
        k - m + \cdots + c - b
    \biggr) 
    \textcolor{lightgray}{+a}
$$

---

# 余数分析

由此可见，数字的这一部分是`11的倍数`
- 因此可被`11整除`

$$
\textcolor{lightgray}{\text{数字} = }
11 \Biggr( 
        kq_{n-1} + 
        mq_{n-2} +
        \cdots +
        cq_{2} + 
        bq_{1}
    \Biggr)
\textcolor{lightgray}{+
\big( \mp k \pm m \mp \cdots \pm c \mp b \big) + a}
$$

因此，要使`整个数字能被11整除`，剩余部分也必须为`11的倍数`

$$
\textcolor{lightgray}{\text{数字} = 
11 \Biggr( 
        kq_{n-1} + 
        mq_{n-2} +
        \cdots +
        cq_{2} + 
        bq_{1}
    \Biggr)} +
\big( \mp k \pm m \mp \cdots \pm c \mp b \big) + a
$$

注意到该表达式包含数字的`所有位`，但采用`交替运算`，除了`a`

$$
\text{数字} \space = \space \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n位}}
$$

## 偶数和奇数情形

考虑两种不同情形下`数字位运算部分能被11整除`的情况

> 当n为`偶数`时：
>
> $$
\underline{- k + m - \cdots + c - b} + a
$$
>
> 当n为`奇数`时：
>
> $$
\underline{k - m + \cdots + c - b} + a
$$

## 极性等价特性

注意到其`负对应式`同样必须能被`11整除`

> 当n为`偶数`时：
>
> $$
\underline{k - m + \cdots - c + b} - a
$$
>
> 当n为`奇数`时：
>
> $$
\underline{- k + m - \cdots - c + b} - a
$$

## 选择正起始形式

因此我们可以选择`以正数开头`的表达式：

> 当n为`偶数`时：
>
> $$
\underline{k - m + \cdots - c + b} - a
$$
>
> 当n为`奇数`时：
>
> $$
\underline{k - m + \cdots + c - b} + a
$$

# 合并为通用表达式

这意味着我们从正数k开始，然后**交替使用正负运算符**

$$
\boxed{
    \phantom{\Big(}
    k - m + \cdots \pm c \mp b \pm a
    \phantom{\Big)}
}
$$

由于求和是交替进行的：
- 位于**奇数位**的数字均为正数
- 位于**偶数位**的数字均为负数

因此，当一个数字表示为：

$$
\text{数字} = \underbrace{\overline{a_{n} \space a_{n-1} \space \cdots \space a_{3} \space a_{2} \space a_{1}}}_{\text{n位}} = 
\space \underline{a_{n} \cdot 10^{n-1}} + 
\underline{a_{n-1} \cdot 10^{n-2}} + 
\cdots + 
\underline{a_{3} \cdot 10^{2}} + 
\underline{a_{2} \cdot 10^{1}} + 
\underline{a_{1} \cdot 10^{0}}
$$

我们可以将所有**奇数位数字**和**偶数位数字**分别组合：

$$
\phantom{\Big(}
    \underline{
        a_1 + a_3 + \cdots + a_{n-3} + a_{n-1}
    } 
    \quad
    \underline{
        - a_2 - a_4 - \cdots - a_{n-2} - a_{n}
    }
\phantom{\Big)}
$$

通过计算**奇数位数字之和**减去**偶数位数字之和**来简化运算：

$$
\boxed{
    \phantom{\Big(}
        \biggr( a_1 + a_3 + \cdots + a_{n-3} + a_{n-1} \biggr)
        - \biggr( a_2 + a_4 + \cdots + a_{n-2} + a_{n} \biggr)
    \phantom{\Big)}
}
$$

# 整除性规则2

因此，要确定 `整个数能否被11整除`，

$$
\boxed{
    \phantom{\biggr(} 
        \textbf{从最大位开始的交替和，必须能被11整除}
    \phantom{\biggr)}
}
$$

$$
\textbf{或者}
$$

$$
\boxed{
    \phantom{\biggr(} 
        \textbf{奇数位置上的数字之和减去偶数位置上的数字之和，必须能被11整除}
    \phantom{\biggr)}
}
$$
