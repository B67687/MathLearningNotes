# 一个数的构成

$$
\text{数} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n位}} =
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

让我们看一下 `10 的幂`

# 分解 10 的幂

我们从 $10^0$ 开始  
- 它等于 `1`

$$
10^{0} = 1
$$

我们无法对它进行 `分割` 或 `因式分解`，因此我们继续看 $10^1$

## 有意分割

注意 `10` 可以以 `11` 和 `1` 表达

$$
\boxed{ 
    \phantom{\big(} 
    10 = \underline{11 - 1} 
    \phantom{\big)} 
}
$$

如果我们继续到下一个 `10 的幂`，我们将乘以 `10` 得到

$$
\boxed{ 
    \phantom{\big(} 
    100 = \underline{11 \cdot 10 - 10} 
    \phantom{\big)} 
}
$$

对于下一个 `10 的幂` 也是一样

$$
\boxed{ 
    \phantom{\big(} 
    1000 = \underline{11 \cdot 10^2 - 10^2} 
    \phantom{\big)} 
}
$$

## 此类分割的一般化

因此，我们得到了一个从 `第一个幂开始` 到某个 `n` 都重复的模式

$$
\begin{align}
    10 &= 11 - 1 \\ 
    100 &= 11 \cdot 10 - 10 \\
    1000 &= 11 \cdot 10^2 - 10^2 \\
    &\vdots \\
    10^n &= 11 \cdot 10^{n-1} - 10^{n-1} \\
\end{align}
$$

# 因式分解这个数

再来看一遍这个数

$$
\text{数} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n位}} = 
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

因为我们知道这个  
- 其中 `n` 从 `1` 开始增加

$$
10^n = 11 \cdot 10^{n-1} - 10^{n-1}
$$

我们将它代入因式分解中

$$
\text{数} = 
\space k \space \biggr( 11 \cdot 10^{n-2} - 10^{n-2} \biggr) + 
m \space \biggr( 11 \cdot 10^{n-3} - 10^{n-3} \biggr) + 
\cdots + 
c \space \biggr( 11 \cdot 10 - 10 \biggr) + 
b \space \biggr( 11 - 1 \biggr) + 
a
$$

简化后，我们得到

$$
\text{数} = 
\boxed{ \phantom{\biggr(} 11 \cdot 10^{n-2}k - 10^{n-2}k \phantom{\biggr)}} +
\boxed{ \phantom{\biggr(} 11 \cdot 10^{n-3}m - 10^{n-3}m \phantom{\biggr)}} + 
\cdots +
\boxed{\phantom{\biggr(} 11 \cdot 10c - 10c \phantom{\biggr)}}  + 
\boxed{\phantom{\biggr(} 11b - b  \phantom{\biggr)}} + 
\boxed{\phantom{\biggr(} a \phantom{\biggr)}}
$$

像这样重新分组

$$
\text{数} = 
\Biggr( 
    11 \cdot 10^{n-2}k + 
    11 \cdot 10^{n-3}m +
    \cdots +
    11 \cdot 10c + 
    11b
\Biggr) -
\biggr( 
    10^{n-2}k + 10^{n-3}m + \cdots + 10c + b
\biggr) +
    a
$$

将这些组按 `11` 重新分组，我们得到

$$
\text{数} = 
11 \Biggr( 
    10^{n-2}k + 
    10^{n-3}m +
    \cdots +
    10c + 
    b
\Biggr) -
\biggr( 
    10^{n-2}k + 10^{n-3}m + \cdots + 10c + b
\biggr) +
a
$$

注意到这也是该数的一部分（在 `十进制` 中）

$$
\text{数} = 
11 \cdot \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} \space -
\underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} +
a
$$


# 余数分析

因此，我们看到数的这一部分是 `11 的倍数`  
- 从而使其 `可被11整除`

$$
\textcolor{lightgray}{\text{数} = }
    11 \cdot \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}}
\space
\textcolor{lightgray}{-}
\space
\textcolor{lightgray}{
    \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} + a
}
$$

因此，要使 `整个数能被11整除`，这一部分也必须是 `11 的倍数`

$$
\textcolor{lightgray}{\text{数} =
    11 \cdot \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}}} \space -
    \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} + a
$$

但它仍然是 `两个表达式`，我们将其简化为 `一个表达式`

$$
\textcolor{lightgray}{\text{数} =
    11 \cdot \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}}} \space -
    \biggr( \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} - a \biggr)
$$

这样我们只需查看 `绝对表达式` 即可

$$
\textcolor{lightgray}{\text{数} =
    11 \cdot \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} \space -}
    \biggr( \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} - a \biggr)
$$

就是这个

$$
\boxed{
    \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} - a
}
$$

# 可被11整除的规则1

因此，要确定 `整个数能否被11整除`，

$$
\textbf{不包括个位的数减去个位上的数，必须能被11整除}
$$
