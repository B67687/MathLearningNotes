# 数字的构成

$$
\text{数字} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n位}} =
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

我们来看看 `10 的幂`

# 10的幂的因式分解

从$10^0$开始
- 等于`1`

$$
10^{0} = 1
$$

无法对此进行"划分"或"因式分解"，因此继续看$10^1$

## 有意划分

注意到`10`可以被划分为包含`3`的部分和剩余部分

$$
\boxed{
    \phantom{\big(}
    10 = 3^2 + 1}
    \phantom{\big)}
$$

对于下一个10的幂，乘以`10`得到：

$$
100 = 3^2 \cdot 10 + \boxed{10}
$$

注意到划分中的`10`，可以像之前那样拆分：

$$
\boxed{
    \phantom{\big(}
    100 = 3^2 \cdot 10 + 3^2 + 1
    \phantom{\big)}    
}
$$

同理，对于下一个幂：

$$
1000 = 3^2 \cdot 100 + 3^2 \cdot 10 + 10
$$

可以表示为：

$$
\boxed{
    \phantom{\big(}
    1000 = 3^2 \cdot 100 + 3^2 \cdot 10 + 3^2 + 1
    \phantom{\big)}
}
$$

## 此类划分的推广

由此得到从`第1个幂`到`n`的重复模式：

$$
\begin{align}
    10 &= 3^2 + 1 \\
    100 &= 3^2 \cdot 10 + 3^2 + 1 \\
    1000 &= 3^2 \cdot 100 + 3^2 \cdot 10 + 3^2 + 1 \\
    &\vdots \\
    10^{n} &= 3^2 \cdot 10^{n-1} + 3^2 \cdot 10^{n-2} + \cdots + 3^2 \cdot 10^{1} + 3^2 \cdot 10^{0} + 1
\end{align}
$$

可以用$3^2$来组合`通用表达式`：

$$
10^{n} = 3^2 \space \Big( 10^{n-1} + 10^{n-2} + \cdots + 10^{1} + 10^{0} \Big) + 1
$$

注意到这实际上是`n位`的每一位`1`：

$$
\textcolor{lightgray}{
    10^{n} = 3^2 \space}
    \Big( 10^{n-1} + 10^{n-2} + \cdots + 10^{1} + 10^{0} \Big)
    \textcolor{lightgray}{+} 
    \textcolor{lightgray}{1}
$$

因此，用数字形式表示为：

$$
\boxed{
    10^{n} = 9 \cdot \underbrace{\overline{11\cdots11}}_{\text{n位}} + 1
}
$$

按`3的倍数`重新分组得到：

$$
\boxed{
    10^{n} = 3 \cdot \underbrace{\overline{33\cdots33}}_{\text{n位}} + 1
}
$$

# 对任意数字的因式分解

回到我们想将任意数字分解为能被 `3` 整除的形式

$$
\text{数字} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n位}} =
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

由于我们知道这个等式成立：

$$
10^{n} = 3 \cdot \underbrace{ \overline{33\cdots33}}_{\text{n位}} + 1
$$

我们将它代入原式中：

$$
\text{数字} = 
k \biggr( 3 \cdot \underbrace{ \overline{33\cdots33}}_{\text{n-1位}} + 1 \biggr) + 
m \biggr( 3 \cdot \underbrace{ \overline{33\cdots33}}_{\text{n-2位}} + 1 \biggr) + 
\cdots + 
c \Biggr( 3 \cdot 33 + 1 \Biggr) + 
b \Biggr( 3 \cdot 3 + 1 \Biggr) + 
a \Biggr(1 \Biggr)
$$

令这些具有重复数字的数表示如下：

$$
q_{n} = \underbrace{\overline{33\cdots33}}_{\text{n位}}
$$

于是我们有：

$$
\text{数字} = 
k \biggr( 3q_{n-1} + 1 \biggr) + 
m \biggr( 3q_{n-2} + 1 \biggr) + 
\cdots + 
c \Biggr( 3q_{2} + 1 \Biggr) + 
b \Biggr( 3q_{1} + 1 \Biggr) + 
a \Biggr(1 \Biggr)
$$

化简后我们得到：

$$
\text{数字} = 
\boxed{ \phantom{\biggr(} 3k \cdot q_{n-1} + k \phantom{\biggr)}} +
\boxed{ \phantom{\biggr(} 3m \cdot q_{n-2} + m \phantom{\biggr)}} + 
\cdots +
\boxed{\phantom{\biggr(} 3c \cdot q_{2} + c \phantom{\biggr)}}  + 
\boxed{\phantom{\biggr(} 3b \cdot q_{1} + b \phantom{\biggr)}} + 
\boxed{\phantom{\biggr(} a \phantom{\biggr)}}
$$

重新整理如下：

$$
\text{数字} = 
\biggr( 
3k \cdot q_{n-1} + 
3m \cdot q_{n-2} +
\cdots +
3c \cdot q_{2} + 
3b \cdot q_{1}
\biggr) +
\big( 
k + m + \cdots + c + b + a
\big)
$$

提取常数因子 `3`，我们得到：

$$
\boxed{
    \text{数字} = 
    3 \biggr( 
    kq_{n-1} + 
    mq_{n-2} +
    \cdots +
    cq_{2} + 
    bq_{1}
    \biggr) +
    \big( 
    k + m + \cdots + c + b + a
    \big)
}
$$


# 余数分析

因此，我们看到该表达式的第一个部分是一个 `3 的倍数`
- 因此它是 `可被 3 整除的`

$$
\textcolor{lightgray}{\text{数字} = }
3 \biggr( 
kq_{n-1} + 
mq_{n-2} +
\cdots +
cq_{2} + 
bq_{1}
\biggr)
\textcolor{lightgray}{+
\big( k + m + \cdots + c + b + a \big)}
$$

因此，要使整个数字 `能被 3 整除`，第二部分也必须是 `3 的倍数`

$$
\textcolor{lightgray}{\text{数字} = 
3 \biggr( 
kq_{n-1} + 
mq_{n-2} +
\cdots +
cq_{2} + 
bq_{1}
\biggr)} +
\big( k + m + \cdots + c + b + a \big)
$$

我们注意到这部分包含了数字本身的所有位数：

$$
\text{数字} \space = \space \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n位}}
$$

# 整除性规则

因此，判断一个 `数字是否能被 3 整除`，只需要：

$$
\boxed{
    \phantom{\biggr(} 
        \textbf{所有位数之和是 3 的倍数}
    \phantom{\biggr)}
}
$$

# 递归迭代

由于一个数字的各位之和总是 `小于或等于数字本身`

---

我们可以先计算原始数字的各位之和，看看是否明显能被 `3` 整除。
- 如果不明显，我们可以继续计算这个和的各位之和，并判断它是否能被 `3` 整除。

> 我们可以不断重复这个操作，直到某一次迭代的结果明显能被 `3` 整除！
