# 数字的构成

$$
\text{数字} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n位}} =
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

我们来看看 `10 的幂`

# 分解 10 的幂

让我们从 $10^0$ 开始  
- 它等于 `1`

$$
10^{0} = 1
$$

我们无法对它进行“分割”或“因式分解”，所以继续看 $10^1$

## 故意进行分割

注意，`10` 可以被分割成一个包含 `3` 的部分和另一个是 `1` 的部分

$$
\boxed{10 = \underline{3^2 + 1}}
$$

如果我们继续到下一个 10 的幂，我们乘以 `10` 得到这个结果

$$
100 = 3^2 \cdot 10 + \boxed{10}
$$

注意到其中的 `10`，我们可以像之前那样把它拆分

$$
\boxed{100 = \underline{3^2 \cdot 10 + 3^2 + 1}}
$$

同样地，当我们继续到下一个幂时，

$$
1000 = 3^2 \cdot 100 + 3^2 \cdot 10 + 10
$$

我们可以这样处理：

$$
\boxed{1000 = \underline{3^2 \cdot 100 + 3^2 \cdot 10 + 3^2 + 1}}
$$

因此，我们得到了一个自我重复的模式

$$
\begin{align}
    10 &= 3^2 + 1 \\
    100 &= 3^2 \cdot 10 + 3^2 + 1 \\
    1000 &= 3^2 \cdot 100 + 3^2 \cdot 10 + 3^2 + 1 \\
    &\vdots \\
\end{align}
$$

## 此种分割方式的广义形式

通过这种模式，我们在每次乘以 `10` 后，把 `10` 分割成 `3` 和 `1`
- 所有非负整数次幂的 `10` 都可以这样分割

$$
10^{n} = \underline{
    3^2 \cdot 10^{n-1} + 3^2 \cdot 10^{n-2} + \cdots + 3^2 \cdot 10^{1} + 3^2 \cdot 10^{0} + 1
}
$$

当我们进一步简化时，得到：

$$
\boxed{
    10^{n} = 3 \cdot \underbrace{\overline{33\cdots33}}_{\text{n位}} + 1
}
$$

我们发现，`10 的幂`可以分成两部分：
- 一部分是可以被 `3` 整除的

$$
\textcolor{lightgray}{10^{n} =} 3 \cdot \underbrace{\overline{33\cdots33}}_{\text{n位}} \textcolor{lightgray}{+ 1}
$$

- 另一部分是 `1`

$$
\textcolor{lightgray}{10^{n} = 3 \cdot \underbrace{\overline{33\cdots33}}_{\text{n位}}} + 1
$$

# 对任意数字的因式分解

回到我们想将任意数字分解为能被 `3` 整除的形式

$$
\text{数字} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n位}} =
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

由于我们知道这个等式成立：

$$
10^{n} = 3 \cdot \underbrace{ \overline{33\cdots33}}_{\text{n位}} + 1
$$

我们将它代入原式中：

$$
\text{数字} = 
k \biggr( 3 \cdot \underbrace{ \overline{33\cdots33}}_{\text{n-1位}} + 1 \biggr) + 
m \biggr( 3 \cdot \underbrace{ \overline{33\cdots33}}_{\text{n-2位}} + 1 \biggr) + 
\cdots + 
c \Biggr( 3 \cdot 33 + 1 \Biggr) + 
b \Biggr( 3 \cdot 3 + 1 \Biggr) + 
a \Biggr(1 \Biggr)
$$

令这些具有重复数字的数表示如下：

$$
q_{n} = \underbrace{\overline{33\cdots33}}_{\text{n位}}
$$

于是我们有：

$$
\text{数字} = 
k \biggr( 3q_{n-1} + 1 \biggr) + 
m \biggr( 3q_{n-2} + 1 \biggr) + 
\cdots + 
c \Biggr( 3q_{2} + 1 \Biggr) + 
b \Biggr( 3q_{1} + 1 \Biggr) + 
a \Biggr(1 \Biggr)
$$

化简后我们得到：

$$
\text{数字} = 
\boxed{ \phantom{\biggr(} 3k \cdot q_{n-1} + k \phantom{\biggr)}} +
\boxed{ \phantom{\biggr(} 3m \cdot q_{n-2} + m \phantom{\biggr)}} + 
\cdots +
\boxed{\phantom{\biggr(} 3c \cdot q_{2} + c \phantom{\biggr)}}  + 
\boxed{\phantom{\biggr(} 3b \cdot q_{1} + b \phantom{\biggr)}} + 
\boxed{\phantom{\biggr(} a \phantom{\biggr)}}
$$

重新整理如下：

$$
\text{数字} = 
\biggr( 
3k \cdot q_{n-1} + 
3m \cdot q_{n-2} +
\cdots +
3c \cdot q_{2} + 
3b \cdot q_{1}
\biggr) +
\big( 
k + m + \cdots + c + b + a
\big)
$$

提取常数因子 `3`，我们得到：

$$
\boxed{
    \text{数字} = 
    3 \biggr( 
    kq_{n-1} + 
    mq_{n-2} +
    \cdots +
    cq_{2} + 
    bq_{1}
    \biggr) +
    \big( 
    k + m + \cdots + c + b + a
    \big)
}
$$


# 结论

因此，我们看到该表达式的第一个部分是一个 `3 的倍数`
- 因此它是 `可被 3 整除的`

$$
\textcolor{lightgray}{\text{数字} = }
3 \biggr( 
kq_{n-1} + 
mq_{n-2} +
\cdots +
cq_{2} + 
bq_{1}
\biggr)
\textcolor{lightgray}{+
\big( k + m + \cdots + c + b + a \big)}
$$

因此，要使整个数字 `能被 3 整除`，第二部分也必须是 `3 的倍数`

$$
\textcolor{lightgray}{\text{数字} = 
3 \biggr( 
kq_{n-1} + 
mq_{n-2} +
\cdots +
cq_{2} + 
bq_{1}
\biggr)} +
\big( k + m + \cdots + c + b + a \big)
$$

我们注意到这部分包含了数字本身的所有位数：

$$
\text{数字} \space = \space \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n位}}
$$

---

因此，判断一个 `数字是否能被 3 整除`，只需要：

$$
\text{所有位数之和是 3 的倍数}
$$

---

# 递归迭代

由于一个数字的各位之和总是 `小于或等于数字本身`

---

我们可以先计算原始数字的各位之和，看看是否明显能被 `3` 整除。
- 如果不明显，我们可以继续计算这个和的各位之和，并判断它是否能被 `3` 整除。

> 我们可以不断重复这个操作，直到某一次迭代的结果明显能被 `3` 整除！

# 示例

让我们用一个例子来验证！

判断这个数字是否能被 `3` 整除：

$$
535137714
$$

我们计算它的各位之和：

$$
\begin{align}
    5 + 3 + 5 + 1 + 3 + 7 + 7 + 1 + 4 &= (5 + 5) + (3 + 7) + (3 + 7) + (1 + 1 + 4) \\
    &= 10 + 10 + 10 + 6 \\
    &= 36
\end{align}
$$

我们知道：

$$
36 = 3 \cdot 12
$$

如果这还不够明显，我们可以再做一次！
- 第二次迭代：

$$
3 + 6 = 9
$$

我们知道 `9` 肯定是 `3 的倍数`

$$
9 = 3 \cdot 3
$$

这是所有能被 `3` 整除的数字在最后一步都会得到的结果！

---

因此，这个数字是 `能被 3 整除的`

$$
535137714 = 3 \cdot 178379238
$$

---

3 \space | \space 535137714   
$$
