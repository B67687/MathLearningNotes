# 数的构成

$$
\text{数} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n位}} =
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

我们来看看“10的幂”

# 10的幂的因式分解

从$10^0$开始
- 等于`1`

$$
10^{0} = 1
$$

无法对此进行"划分"或"因式分解"，因此继续看$10^1$

## 有意划分

注意到`10`可以划分为`包含4的商`和`余数2`

$$
\boxed{ 
    \phantom{\big(} 
    10 = 8 + 2 
    \phantom{\big)} 
}
$$

对于下一个10的幂，乘以`10`得到：

$$
100 = 8 \cdot 10 + 2 \cdot \boxed{10}
$$

注意到划分中的`10`，可以像之前那样拆分：

$$
\begin{align}
    100 &= 8 \cdot 10 + \underline{2 \space \big( 8 + 2 \big)} \\
    &= 8 \cdot 10 + 8 \cdot 2 + 2^2 \\
\end{align}
$$

得到：

$$
\boxed{
    \phantom{\big(} 
    100 =  8 \cdot 10 + 8 \cdot 2 + 2^2 
    \phantom{\big)}
}
$$

对于下一个幂：

$$
1000 = 8 \cdot 10^2 + 8 \cdot 2 \cdot 10 + 2^2 \cdot \boxed{10} 
$$

再次拆分：

$$
\begin{align}
    1000 &= 8 \cdot 10^2 + 8 \cdot 2 \cdot 10 + \underline{2^2 \space (8 + 2)} \\
    &= 8 \cdot 10^2  + 8 \cdot 2 \cdot 10 + 8 \cdot 2^2 + 2^3
\end{align}
$$

因此：

$$
\boxed{
    \phantom{\big(} 
    1000 = 8 \cdot 10^2 + 8 \cdot 2 \cdot 10 + 8 \cdot 2^2 + 2^3
    \phantom{\big)}
}
$$

对于下一个幂：

$$
10000 = 8 \cdot 10^3 + 8 \cdot 2 \cdot 10^2 + 8 \cdot 2^2 \cdot 10 + 2^3 \cdot \boxed{10}
$$

再次拆分：

$$
\begin{align}
    10000 &= 8 \cdot 10^3 + 8 \cdot 2 \cdot 10^2 + 8 \cdot 2^2 \cdot 10 + \underline{2^3 \space (8 + 2)} \\
    &= 8 \cdot 10^3 + 8 \cdot 2 \cdot 10^2 + 8 \cdot 2^2 \cdot 10 + 8 \cdot 2^3 + 2^4
\end{align}
$$

因此：

$$
\boxed{
    \phantom{\big(} 
    10000 = 8 \cdot 10^3 + 8 \cdot 2 \cdot 10^2 + 8 \cdot 2^2 \cdot 10 + 8 \cdot 2^3 + 2^4
    \phantom{\big)}
}
$$

## 此类划分的推广

由此得到从`第3个幂`到`n`的重复模式：

$$
\begin{align}
    10 &= 8 + 2 \\
    100 &= 8 \cdot 10 + 8 \cdot 2 + 2^2 \\
    1000 &= 8 \cdot 10^2 + 8 \cdot 2 \cdot 10 + 8 \cdot 2^2 + 2^3 \\
    10000 &= 8 \cdot 10^3 + 8 \cdot 2 \cdot 10^2 + 8 \cdot 2^2 \cdot 10 + 8 \cdot 2^3 + 2^4 \\
    &\vdots \\
    10^n &= 8 \cdot 2^{0} \cdot 10^{n-1} + 8 \cdot 2^{1} \cdot 10^{n-2} + \cdots + 8 \cdot 2^{n-2} \cdot 10^{1} + 8 \cdot 2^{0} \cdot 10^{0} + 2^{n} \\
\end{align}
$$

因此`通用表达式`为：

$$
\boxed{
    \phantom{\Biggr(}
        10^{n} = 8 \biggr( 2^{0} \cdot 10^{n-1} + 2^{1} \cdot 10^{n-2} + \cdots + 2^{n-2} \cdot 10^{1} + 2^{n-1} \cdot 10^{0} \biggr) + 2^{n} 
    \phantom{\Biggr)}
}
$$

# 覆盖大多数情况的特殊条件

注意到我们含有$2^n$项

$$
\textcolor{lightgray}{
    8 + 
    \biggr( 2^{0} \cdot 10^{n-1} + 2^{1} \cdot 10^{n-2} + \cdots + 2^{n-2} \cdot 10^{1} + 2^{0} \cdot 10^{0} \biggr)
} + 2^{n}
$$

而`8`也可以表示为$2^3$

$$
2^3 
\textcolor{lightgray}{
    \biggr( 2^{0} \cdot 10^{n-1} + 2^{1} \cdot 10^{n-2} + \cdots + 2^{n-2} \cdot 10^{1} + 2^{0} \cdot 10^{0} \biggr)
} + 2^{n}
$$

因此，当$n \ge 3$时，我们可以这样做：

$$
2^3 \space \biggr( 2^{0} \cdot 10^{n-1} + 2^{1} \cdot 10^{n-2} + \cdots + 2^{n-2} \cdot 10^{1} + 2^{0} \cdot 10^{0} \biggr) + 2^{3} \cdot 2^{n-3}
$$

这样就能将`2的幂表达式`也归入`8的倍数`部分
> 使其成为单一的`8的倍数`表达式

$$
2^3 \space \biggr( 2^{0} \cdot 10^{n-1} + 2^{1} \cdot 10^{n-2} + \cdots + 2^{n-2} \cdot 10^{1} + 2^{0} \cdot 10^{0} + 2^{n-3} \biggr)
$$

因此，对于$n \ge 3$的一般表达式为：

$$
\boxed{
    \phantom{\Biggr(}
    10^{n} = 8 \biggr( 2^{0} \cdot 10^{n-1} + 2^{1} \cdot 10^{n-2} + \cdots + 2^{n-2} \cdot 10^{1} + 2^{n-1} \cdot 10^{0} + 2^{n-3} \biggr)
    \phantom{\Biggr)}
}
$$

# 通用表达式的应用

我们可以将这个表达式应用于几乎所有的`10的幂`

> 只要满足$n \ge 3$

$$
10^{n} = 8 \biggr( 2^{0} \cdot 10^{n-1} + 2^{1} \cdot 10^{n-2} + \cdots + 2^{n-2} \cdot 10^{1} + 2^{n-1} \cdot 10^{0} + 2^{n-3} \biggr)
$$

而对于$0 \le n \le 2$的一般表达式为：

$$
10^{n} = 8 \biggr( 2^{0} \cdot 10^{n-1} + 2^{1} \cdot 10^{n-2} + \cdots + 2^{n-2} \cdot 10^{1} + 2^{n-1} \cdot 10^{0} \biggr) + 2^{n} 
$$

这实际上可以简化为以下`2个简单表达式`：

$$
\boxed{
    \phantom{\big(}
    10^1 = 8 + 2
    \phantom{\big)}
}
$$

$$
\boxed{
    \phantom{\big(}
    10^2 = 8 \cdot 12 + 2^2
    \phantom{\big)}
}
$$

# 按照结构来看

我们再回过头来看看这个数的结构：

$$
\text{数} = \underbrace{\overline{k \space m \space \cdots \space d \space c \space b \space a}}_{\text{n位}} = 
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{d \cdot 10^{3}} +
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

## 按位可除性分析

因为我们知道从第三幂开始的所有 `10的幂` 都是 `8的倍数`

$$
\text{数} = \underbrace{\overline{k \space m \space \cdots \space d \space \textcolor{lightgray}{c \space b \space a}}}_{\text{n位}} = 
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{d \cdot 10^{3}} \textcolor{lightgray}{+
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}}
$$

所以我们只需要检查最后三位（即百位、十位和个位组成的数）是否能被 `8` 整除即可。

$$
\text{数} = \underbrace{\overline{\textcolor{lightgray}{k \space m \space \cdots \space d} \space c \space b \space a}}_{\text{n位}} = 
\textcolor{lightgray}{\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{d \cdot 10^{3}}} +
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

# 可除性规则 1

因此，要判断整个数是否能被 `8` 整除，  

$$
\boxed{
    \phantom{\biggr(} 
        \textbf{最后三位组成的数必须是8的倍数}
    \phantom{\biggr)}
}
$$

# 数字的因式分解

重新观察数字表达式：

$$
\text{数字} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n位}} =
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{d \cdot 10^{3}} + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

已知：

> 对于$n \ge 3$
>
> $$
10^{n} = 8 \biggr( 2^{0} \cdot 10^{n-1} + 2^{1} \cdot 10^{n-2} + \cdots + 2^{n-2} \cdot 10^{1} + 2^{n-1} \cdot 10^{0} + 2^{n-3} \biggr)
$$

以及

> 对于$0 \le n \le 2$
>
> $$
\boxed{
    \phantom{\big(}
    10^1 = 8 + 2
    \phantom{\big)}
}
$$
>
> $$
\boxed{
    \phantom{\big(}
    10^2 = 8 \cdot 12 + 2^2
    \phantom{\big)}
}
$$

## 大表达式简化

由于这部分已经是`8的倍数`：

$$
\textcolor{lightgray}{10^n =} 8 \biggr( 2^{0} \cdot 10^{n-1} + 2^{1} \cdot 10^{n-2} + \cdots + 2^{n-2} \cdot 10^{1} + 2^{n-1} \cdot 10^{0} \biggr)
$$

设$q_{n} = 2^{0} \cdot 10^{n-1} + 2^{1} \cdot 10^{n-2} + \cdots + 2^{n-2} \cdot 10^{1} + 2^{n-1} \cdot 10^{0}$

> 对于$n \ge 3$
>
> $$
\boxed{
    \phantom{\big(}
    10^n = 8q_{n}
    \phantom{\big)}
}
$$

## 代数运算

将幂表达式代入数字分解：

$$
\text{数字} = 
k \space \biggr( 8q_{n-1} \biggr) + 
m \space \biggr( 8q_{n-2} \biggr) + 
\cdots + 
d \space \biggr( 8q_{3} \biggr) + 
c \space \biggr( 8 \cdot 12 + 2^{2} \biggr) + 
b \space \biggr( 8 + 2 \biggr) + 
a
$$

简化后得到：

$$
\text{数字} = 
\boxed{ \phantom{\biggr(} 8k \cdot q_{n-1} \phantom{\biggr)}} +
\boxed{ \phantom{\biggr(} 8m \cdot q_{n-2} \phantom{\biggr)}} + 
\cdots +
\boxed{\phantom{\biggr(} 8d \cdot q_{3} \phantom{\biggr)}}  + 
\boxed{\phantom{\biggr(} 8 \cdot 12c + 4c \phantom{\biggr)}}  + 
\boxed{\phantom{\biggr(} 8b + 2b  \phantom{\biggr)}} + 
\boxed{\phantom{\biggr(} a \phantom{\biggr)}}
$$

重新分组：

$$
\text{数字} = 
\Biggr( 
    8 \cdot kq_{n-1} + 
    8 \cdot mq_{n-2} +
    \cdots +
    8 \cdot dq_{3} +
    96c +
    8b
\Biggr) +
\biggr( 
    4c +
    2b +
    a
\biggr)
$$

按8合并后：

$$
\text{数字} = 
8 \Biggr( 
    kq_{n-1} + 
    mq_{n-2} +
    \cdots +
    dq_{3} +
    12c +
    b
\Biggr) +
\biggr( 
    4c +
    2b +
    a
\biggr)
$$

简化为：

$$
\boxed{
    \text{数字} = 
    8 \Biggr( 
        kq_{n-1} + 
        mq_{n-2} +
        \cdots +
        dq_{3} +
        12c +
        b
    \Biggr) +
    \biggr( 
        4c +
        2b +
        a
    \biggr)
}
$$

# 余数分析

由此可见，数字的这一部分是`8的倍数`：

$$
\textcolor{lightgray}{\text{数字} =} 
    8 \Biggr( 
        kq_{n-1} + 
        mq_{n-2} +
        \cdots +
        dq_{3} +
        12c +
        b
    \Biggr) 
    \textcolor{lightgray}{+
        \biggr(
            4c +
            2b +
            a
        \biggr)
    }
$$

因此，要使`整个数字能被8整除`，剩余部分也必须为`8的倍数`：

$$
\textcolor{lightgray}{\text{数字} = 
   8 \Biggr( 
        kq_{n-1} + 
        mq_{n-2} +
        \cdots +
        dq_{3} +
        12c +
        b
    \Biggr) 
    } 
    \textcolor{lightgray}{+}
    \textcolor{lightgray}{\biggr(}
    4c +
    2b +
    a
    \textcolor{lightgray}{\biggr)}
$$

这实际上是**百位数乘以4**加上**十位数乘以2**再加上**个位数**的和：

$$
\boxed{
    \phantom{\big(}
    4c + 2b + a
    \phantom{\big)}
}
$$

# 可除性规则 2

因此，要判断整个数是否能被 `8` 整除，  

$$
\boxed{
    \phantom{\biggr(} 
        \textbf{该数: } \underline{\textbf{倒数第三位的四倍}} \textbf{加上 } \underline{\textbf{倒数第二位的两倍}} \textbf{, 再加上 } \underline{\textbf{最后一位数字}} \textbf{，必须能被8整除}
    \phantom{\biggr)}
}
$$
