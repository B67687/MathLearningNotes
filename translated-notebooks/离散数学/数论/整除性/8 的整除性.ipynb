# 数的构成

$$
\text{数} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n位}} =
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

我们来看看“10的幂”

# 对10的幂进行因式分解

从 $10^0$ 开始  
- 它等于 `1`

$$
10^{0} = 1
$$

这个数既不能“拆分”也不能“因式分解”，所以我们继续看下一个，即 $10^1$

## 有意识的拆分

注意到 `10` 可以被“拆分成”一个包含 `8` 的商和余数为 `2` 的形式：

$$
\boxed{ \phantom{\big(} 10 = \underline{8 + 2} \phantom{\big)} }
$$

如果我们继续到下一个10的幂，我们将前面的结果乘以 `10` 得到：

$$
100 = 8 \cdot 10 + 2 \cdot \boxed{10}
$$

注意到了其中的 `10`，我们可以像之前那样对它进一步拆分：

$$
100 = 8 \cdot 10 + \underline{2 \space \big( 8 + 2 \big)}
$$

然后按 `8` 进行重组：

$$
\begin{align}
    100 &= 8 \cdot 10  + 8 \cdot 2 + 2 \cdot 2 \\
    &= 8 \space (10 + 2) + 2 \cdot 2 \\
    &= 8 \cdot 12 + 2 \cdot 2 \\
\end{align}
$$

于是得到：

$$
\boxed{
    100 = \phantom{\big(} 8 \cdot 12 + 2 \cdot 2 \phantom{\big)}
}
$$

当我们继续到下一个幂次时，

$$
1000 = 8 \cdot 120 + \underline{2 \cdot 20}
$$

按 `8` 重新组合：

$$
\begin{align}
    1000 &= \textcolor{lightgray}{8 \cdot 120} + 2 \cdot 2 \space (8 + 2) \\
    &= \textcolor{lightgray}{8 \cdot 120} + 4 \space (8 + 2) \\
    &= \textcolor{lightgray}{8 \cdot 120} + 8 \cdot 4 + 4 \cdot 2 \\
    &= \textcolor{lightgray}{8 \cdot 120} + 8 \cdot 4 + \underline{8 \cdot 1} \\
    &= 8 \space (100 + 20 + 4 + 1) \\
    &= 8 \cdot 125
\end{align}
$$

因此，

$$
\boxed{
    \phantom{\big(} 1000 = \underline{8 \cdot 125} \phantom{\big)}
}
$$

这种模式在接下来的幂次中依然成立：

$$
\boxed{ \phantom{\big(} 10000 = \underline{8 \cdot 1250} \phantom{\big(}}
$$

再下一项：

$$
\boxed{ \phantom{\big(} 100000 = \underline{8 \cdot 12500} \phantom{\big(}}
$$

## 这种拆分方式的一般化表达

于是我们得到了如下从第三幂开始一直到某个 `n` 的规律性模式：

$$
\begin{align}
    1000 &= 8 \cdot 125 \\
    10000 &= 8 \cdot 1250 \\
    100000 &= 8 \cdot 12500 \\
    &\vdots \\
    10^n &= 8 \space \big( 125 \cdot 10^{n-3} \big)
\end{align}
$$

# 按照结构来看

我们再回过头来看看这个数的结构：

$$
\text{数} = \underbrace{\overline{k \space m \space \cdots \space d \space c \space b \space a}}_{\text{n位}} = 
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{d \cdot 10^{3}} +
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

## 按位可除性分析

因为我们知道从第三幂开始的所有 `10的幂` 都是 `8的倍数`

$$
\text{数} = \underbrace{\overline{k \space m \space \cdots \space d \space \textcolor{lightgray}{c \space b \space a}}}_{\text{n位}} = 
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{d \cdot 10^{3}} \textcolor{lightgray}{+
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}}
$$

所以我们只需要检查最后三位（即百位、十位和个位组成的数）是否能被 `8` 整除即可。

$$
\text{数} = \underbrace{\overline{\textcolor{lightgray}{k \space m \space \cdots \space d} \space c \space b \space a}}_{\text{n位}} = 
\textcolor{lightgray}{\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{d \cdot 10^{3}}} +
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

# 可除性规则 1

因此，要判断整个数是否能被 `8` 整除，  

$$
\boxed{
    \phantom{\biggr(} 
        \textbf{最后三位组成的数必须是8的倍数}
    \phantom{\biggr)}
}
$$

---

现在对于大数来说这可能不太容易判断，所以也许我们可以通过**因式分解**的方法来观察是否存在另一种判断方式。

---

# 对数进行因式分解

我们再次看看这个数：

$$
\text{数} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n位}} =
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{d \cdot 10^{3}} + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

我们知道以下关系成立：
- 其中 `n ≥ 3`

$$
10^{n} = 8 \space \big( 125 \cdot 10^{n-3} \big)
$$

我们将它代入原式中：

$$
\text{数} = 
k \cdot \boxed{8 \space \big( 125 \cdot 10^{n-4} \big)} + 
m \cdot \boxed{8 \space \big( 125 \cdot 10^{n-5} \big)} + 
\cdots + 
d \cdot \boxed{8 \space \big( 125 \cdot 10^0 \big)} + 
c \cdot \boxed{8 \space \big( 12 \big) + 2 \big( 2 \big)} + 
b \cdot \boxed{8 + 2} + 
a
$$

将所有的 `125 × 10的幂` 表示为：

$$
q_{n} = 125 \cdot 10^{n} 
$$

于是我们得到：

$$
\text{数} = 
8k \cdot \underline{q_{n-4}} + 
8m \cdot \underline{q_{n-5}} + 
\cdots + 
8d \cdot \underline{q_{0}} + 
\boxed{
    \phantom{\big(} 
    8c \cdot \underline{12} + 4c
    \phantom{\big)}
} +
\boxed{
    \phantom{\big(}
    8b \cdot \underline{1} + 2b
    \phantom{\big)}
} +
a
$$

重新整理一下：

$$
\text{数} = 
    8 \underline{kq_{n-4}} + 
    8 \underline{mq_{n-5}} + 
    \cdots + 
    8 \underline{dq_{0}} + 
    8 \space \underline{\big( 12c \big)} +
    8 \underline{b} + 
    \boxed{
        4c +
        2b +
        a
    }
$$

提取出公共因子 `8` 后得到：

$$
\boxed{
    \text{数} = 
    8 \biggr( 
    kq_{n-4} + 
    mq_{n-5} +
    \cdots +
    dq_{0} +
    12c +
    b
    \biggr) +
    4c +
    2b +
    a
}
$$

# 余数分析

因此，我们看到这个数的一部分是一个 `8` 的倍数  
- 所以它是可以被 `8` 整除的部分

$$
\textcolor{lightgray}{\text{数} =} 
    8 \biggr( 
    kq_{n-4} + 
    mq_{n-5} +
    \cdots +
    dq_{0} +
    12c +
    b
    \biggr) \textcolor{lightgray}{+
    4c +
    2b +
    a}
$$

所以为了让整个数能被 `8` 整除，它的“余数”也必须是 `8` 的倍数：

$$
\textcolor{lightgray}{\text{数} = 
    8 \biggr( 
    kq_{n-4} + 
    mq_{n-5} +
    \cdots +
    dq_{0} +
    12c +
    b
    \biggr)} +
    4c +
    2b +
    a
$$


# 可除性规则 2

因此，要判断整个数是否能被 `8` 整除，  

$$
\boxed{
    \phantom{\biggr(} 
        \textbf{该数: } \underline{\textbf{倒数第三位的四倍}} \textbf{加上 } \underline{\textbf{倒数第二位的两倍}} \textbf{, 再加上 } \underline{\textbf{最后一位数字}} \textbf{，必须能被8整除}
    \phantom{\biggr)}
}
$$
