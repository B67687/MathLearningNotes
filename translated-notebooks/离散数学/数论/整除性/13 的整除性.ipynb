# 数字的构成

$$
\text{数字} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n位}} =
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

让我们观察“10的幂”

# 分解10的幂

从$10^0$开始
- 等于`1`

$$
10^{0} = 1
$$

无法对此进行“划分”或“分解”，因此继续看$10^1$

## 有意划分

注意到`13`可以表示为`10`和`3`

$$
\boxed{ 
    \phantom{\big(} 
        10 = 13 - 3 
    \phantom{\big)} 
}
$$

对于“下一个10的幂”，我们乘以`10`得到

$$
\boxed{ 
    \phantom{\big(} 
    100 = 13 \cdot 10 - 3 \cdot 10 
    \phantom{\big)} 
}
$$

同理，对于“再下一个10的幂”

$$
\boxed{ 
    \phantom{\big(} 
    1000 = 13 \cdot 10^2 - 3 \cdot 10^2
    \phantom{\big)} 
}
$$

## 此类划分的推广

由此得到从“第1个幂”到某个“n”的重复模式

$$
\begin{align}
    10 &= 13 - 3 \\ 
    100 &= 13 \cdot 10 - 3 \cdot 10 \\
    1000 &= 13 \cdot 10^2 - 3 \cdot 10^2 \\
    &\vdots \\
    10^n &= 13 \cdot 10^{n-1} - 3 \cdot 10^{n-1} \\
\end{align}
$$

因此通用表达式为

$$
\boxed{
    \phantom{\big(}
        10^n = 13 \cdot 10^{n-1} - 3 \cdot 10^{n-1}
    \phantom{\big)}
}
$$



# 数字的因式分解

回顾数字表达式

$$
\text{数字} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n位}} = 
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

已知（`n`从`1`开始）

$$
10^n = 13 \cdot 10^{n-1} - 3 \cdot 10^{n-1}
$$

将其代入因式分解

$$
\text{数字} = 
\space k \space \biggr( 13 \cdot 10^{n-2} - 3 \cdot 10^{n-2} \biggr) + 
m \space \biggr( 13 \cdot 10^{n-3} - 3 \cdot 10^{n-3} \biggr) + 
\cdots + 
c \space \biggr( 13 \cdot 10 - 3 \cdot 10 \biggr) + 
b \space \biggr( 13 - 3 \biggr) + 
a
$$

简化后得到

$$
\text{数字} = 
\boxed{ \phantom{\biggr(} 13 \cdot 10^{n-2}k - 3 \cdot 10^{n-2}k \phantom{\biggr)}} +
\boxed{ \phantom{\biggr(} 13 \cdot 10^{n-3}m - 3 \cdot 10^{n-3}m \phantom{\biggr)}} + 
\cdots +
\boxed{\phantom{\biggr(} 13 \cdot 10c - 3 \cdot 10c \phantom{\biggr)}}  + 
\boxed{\phantom{\biggr(} 13b - 3b  \phantom{\biggr)}} + 
\boxed{\phantom{\biggr(} a \phantom{\biggr)}}
$$

重新分组如下

$$
\text{数字} = 
\Biggr( 
    13 \cdot 10^{n-2}k + 
    13 \cdot 10^{n-3}m +
    \cdots +
    13 \cdot 10c + 
    13b
\Biggr) -
\biggr( 
    3 \cdot 10^{n-2}k +
    3 \cdot 10^{n-3}m +
    \cdots +
    3 \cdot 10c + 
    3b
\biggr) +
    a
$$

按`13`和`3`重新分组

$$
\text{数字} = 
13 \Biggr( 
    10^{n-2}k + 
    10^{n-3}m +
    \cdots +
    10c + 
    b
\Biggr) -
3 \biggr( 
    10^{n-2}k +
    10^{n-3}m +
    \cdots +
    10c +
    b
\biggr) +
a
$$

可识别为`n-1位`数字的`基10表示`

$$
\text{数字} = 
13 \cdot \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} \space -
3 \cdot \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} +
a
$$


# 余数分析

可见这部分是`13的倍数`
- 因此可被`13整除`

$$
\textcolor{lightgray}{\text{数字} = }
    13 \cdot \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}}
\space
\textcolor{lightgray}{-}
\space
\textcolor{lightgray}{
    3 \cdot \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} + a
}
$$

因此，若`整个数字能被13整除`，则剩余部分也必须是`13的倍数`

$$
\textcolor{lightgray}{\text{数字} =
    13 \cdot \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}}} \space -
    3 \cdot \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} + a
$$

将其简化为单一表达式

$$
\textcolor{lightgray}{\text{数字} =
    13 \cdot \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}}} \space -
    \biggr( 3 \cdot \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} - a \biggr)
$$

最终得到核心判断式

$$
\boxed{
    3 \cdot \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} - a
}
$$

# 可除性规则1

因此，判断`整个数字能否被13整除`的条件为：

$$
\boxed{
    \phantom{\biggr(} 
        \underline{
            \textbf{去掉个位数后剩余数字的3倍}
        } 
        \space 
        \textbf{减去} 
        \space 
        \underline{\textbf{个位数字}} 
        \space 
        \textbf{的结果必须能被13整除}
    \phantom{\biggr)}
}
$$

# 寻找更优规则

若将`a`整合到`13的倍数`部分，使得只需关注`括号内表达式`会如何？

$$
\textcolor{lightgray}{\text{数字} = 
13 \cdot \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} -
3 \cdot \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} \space +} \space
a
$$

为实现完全整合，需将`a`拆分为两部分：
- 一部分并入`13的倍数`表达式
- 另一部分并入`3的倍数`表达式

如下所示

$$
\textcolor{lightgray}{\text{数字} = 
13 \cdot \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} -
3 \cdot \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} \space +} \space
(13a - 3 \cdot 4a)
$$

整合后表达式

$$
\textcolor{lightgray}{\text{数字} = 
13 \Biggr( 
    \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}}} + a 
\textcolor{lightgray}{\Biggr)}
\textcolor{lightgray}{-
3 \biggr( 
    \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}}} + 4a 
\textcolor{lightgray}{\biggr)}
$$

此时只需检验括号内表达式是否能被`13`整除

$$
\textcolor{lightgray}{\text{数字} = 
13 \Biggr( 
   \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} + a
\Biggr) -
3 \biggr(} 
    \underbrace{\overline{k \space m \space \cdots \space c \space b}}_{\text{n-1位}} + 4a 
\textcolor{lightgray}{\biggr)}
$$

可见该表达式包含数字的`前n-1位`与`4a`的组合

$$
\boxed{
    \underbrace{\overline{k \space m \space \cdots \space c \space b \space}}_{\text{n-1位}} + 4a
}
$$


# 可除性规则2

因此判断`整个数字能否被13整除`的条件为：

$$
\boxed{
    \phantom{\biggr(} 
        \underline{
            \textbf{去掉个位数后的数字}
        } 
        \space 
        \textbf{加上} 
        \space 
        \underline{\textbf{个位数字的4倍}} 
        \space 
        \textbf{必须能被13整除}
    \phantom{\biggr)}
}
$$

# 深度划分

注意到`10`可表示为`13`与`3`的组合

$$
\boxed{ 
    \phantom{\big(} 
    10 = 13 - 3
    \phantom{\big)} 
}
$$

对于`下一个10的幂`，乘以`10`得到

$$ 
100 = 13 \cdot 10 - 3 \cdot \boxed{10}
$$

其中`10`是唯一非`13`的表达式
- 可继续拆分

$$ 
\begin{align}
    100 &= \textcolor{lightgray}{13 \cdot 10} - 3 \cdot (13 - 3) \\
    &= 13 \cdot 10 - 13 \cdot 3 + 3 \cdot 3 \\
    &= 13 \cdot 10 - 13 \cdot 3 + 3^2 \\
\end{align}
$$

最终得到

$$
\boxed{ 
    \phantom{\big(} 
        100 = 13 \cdot 10 - 13 \cdot 3 + 3^2
    \phantom{\big)} 
}
$$

对于`再下一个10的幂`：

$$
1000 = 13 \cdot 10^2 - 13 \cdot 3 \cdot 10 + 3^2 \cdot \boxed{10}
$$

同样拆分`10`：

$$
\begin{align}
    1000 &= 13 \cdot 10^2 - 13 \cdot 3 \cdot 10 + 3^2 \cdot (13 - 3) \\
    &= 13 \cdot 10^2 - 13 \cdot 3 \cdot 10 + 13 \cdot 3^2 - 3^3 \\
\end{align}
$$

结果表达式：

$$
\boxed{ 
    \phantom{\big(} 
    1000 = 13 \cdot 10^2 - 13 \cdot 3 \cdot 10 + 13 \cdot 3^2 - 3^3 
    \phantom{\big)} 
}
$$

## 划分模式的推广

由此得到从`第1个幂`到`n`的通用模式：

$$
\begin{align}
    10 &= 13 - 3 \\ 
    100 &= 13 \cdot 10 - 13 \cdot 3 + 3^2 \\
    1000 &= 13 \cdot 10^2 - 13 \cdot 3 \cdot 10 + 13 \cdot 3^2 - 3^3 \\
    &\vdots \\
        10^n &= 13 \cdot 3^0 \cdot 10^{n-1} - 13 \cdot 3^1 \cdot 10^{n-2} + \cdots \pm 13 \cdot 3^{n-1} \cdot 10^0 \mp 3^n \\
\end{align}
$$

简化后的通用表达式：

- **统一表达式**

$$
\boxed{
    10^n = 13 \biggr( 3^0 \cdot 10^{n-1} - 3^1 \cdot 10^{n-2} + \cdots \pm 3^{n-1} \cdot 10^{0} \biggr) \mp 3^{n}
}
$$

---

> 当**n为奇数**时
>
> $$
\textcolor{lightgray}{
    10^n = 13 \biggr( 3^0 \cdot 10^{n-1} - 3^1 \cdot 10^{n-2} + \cdots
    } 
    + 3^{n-1} \cdot 10^{0} 
    \textcolor{lightgray}{\biggr)} 
    - 3^{n}
$$
>
> 当**n为偶数**时
>
>$$
\textcolor{lightgray}{
    10^n = 13 \biggr( 3^0 \cdot 10^{n-1} - 3^1 \cdot 10^{n-2} + \cdots
    }
    - 3^{n-1} \cdot 10^{0} 
    \textcolor{lightgray}{\biggr)}
    + 3^{n}
$$

# 数字的因式分解

重新观察数字表达式

$$
\text{数字} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n位}} = 
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

已知当`n`从`1`开始时的幂表达式：

$$
10^n = 13 \biggr( 3^0 \cdot 10^{n-1} - 3^1 \cdot 10^{n-2} + \cdots \pm 3^{n-1} \cdot 10^{0} \biggr) \mp 3^{n}
$$

## 表达式简化

由于该部分已是`13的倍数`：

$$
\textcolor{lightgray}{10^n =} 13 \biggr( 3^0 \cdot 10^{n-1} - 3^1 \cdot 10^{n-2} + \cdots \pm 3^{n-1} \cdot 10^{0} \biggr) \textcolor{lightgray}{\mp} \textcolor{lightgray}{3^{n}}
$$

设$q_{n} = 3^0 \cdot 10^{n-1} - 3^1 \cdot 10^{n-2} + \cdots \pm 3^{n-1} \cdot 10^{0}$，则可简化为：

$$
\boxed{
    10^n = 13q_{n} \mp 3^n
}
$$

---

>当n为`奇数`时：
>
>$$
\textcolor{lightgray}{
    10^n = 13q_{n}
} - 3^n
$$
>
>当n为`偶数`时：
>
> $$
\textcolor{lightgray}{
    10^n = 13q_{n}
} + 3^n
$$

---

$$
\textcolor{lightgray}{\text{数字} =
