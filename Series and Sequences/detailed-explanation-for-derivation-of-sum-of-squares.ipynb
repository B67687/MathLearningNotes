# Combinatorial Approach to the Sum of Squares Formula

We derive the formula for the sum of squares using a combinatorial approach by expressing the sum as a double summation, applying the arithmetic sequence formula, and then using algebraic manipulation.

---

## Transforming the Sum into a Familiar Form

With a bit of play, we can break the sum of $\sum_{k=1}^n k^2$ like this:

$$
\sum_{k=1}^n k^2 = 1^2 + 2^2 + 3^2 + ... + n^2
$$

$$
\sum_{k=1}^n k^2 = (1 \cdot 1) + (2 \cdot 2) + (3 \cdot 3) + ... + (n \cdot n)
$$

$$
\sum_{k=1}^n k^2 = 1 + (2 + 2) + (3 + 3 + 3) + ... + (\underbrace{n + n + \dots + n}_{n \text{ times}})
$$

$$
\sum_{k=1}^n k^2 = \underbrace{(1 + 2 + 3 + ... + n) + (2 + 3 + 4 + ... + n) + ... + n}_{n \text{ terms}}
$$

$$
\sum_{k=1}^n k^2 = \underbrace{\sum_{j=1}^n j + \sum_{j=2}^n j + ... + \sum_{j=n}^n j}_{n \text{ terms}}
$$

And thus we get this Double Summation:

$$
\boxed{\sum_{k=1}^n k^2 = \sum_{k=1}^n \sum_{j=k}^{n} j}
$$

---

## Visualising the Sum in 2 Dimensions

We can also visualise this transformation much more distinctly by setting the terms up in a 2-dimensional form:

$$
\begin{array}{c|ccccc}
    & j=1 & j=2 & j=3 & \cdots & \cdots & \cdots & j=n \\
\hline
    k=1 & 1 \\
    k=2 & 2 & 2 \\
    k=3 & 3 & 3 & 3 \\
    k=4 & 4 & 4 & 4 & 4 \\
    k=5 & 5 & 5 & 5 & 5 & 5 \\
    \vdots & \vdots & \vdots & \vdots & \vdots & \vdots & \ddots \\
    k=n & n & n & n & n & n & \cdots & n \\
\hline
    \text{Sum} & \sum_{i=1}^{n} i & \sum_{i=2}^{n} i & \sum_{i=3}^{n} i & \sum_{i=4}^{n} i & \sum_{i=5}^{n} i & \cdots & \sum_{i=n}^{n} i
\end{array}
$$

By arranging it like this, we can realise that although we can sum it the way we were told, left-right then top-down, it can be expressed equally as well in Summation Notation:

$$
\sum_{\text{row } k=1}^{n} (\text{sum of row } k) = \sum_{k=1}^{n} \sum_{m=1}^{j} j = \sum_{k=1}^{n} (j \times j) = \sum_{k=1}^{n} i^2
$$

We can notice that we can also sum it in terms of arithmetic sequences as seen if we choose to follow the sum top-down then left-right.

$$
\sum_{k=1}^n k^2 = \sum_{k=1}^{n} \sum_{j=k}^{n} j
$$

And thus we get this Double Summation:

$$
\boxed{\sum_{k=1}^n k^2 = \sum_{k=1}^{n} \sum_{j=k}^{n} j}
$$

## Apply the Arithmetic Sequence Formula

Let
$$
\mathbf{S} = \sum_{k=1}^n k^2 = \sum_{k=1}^{n} \sum_{j=k}^{n} j
$$

The inner sum $\sum_{i=k}^n i$ is an arithmetic sequence with:
- First term: $k$
- Last term: $n$
- Number of terms: $n - k + 1$

The sum of an arithmetic sequence is:

$$
\left( 
    \dfrac{\text{first term} + \text{last term}}{2} 
\right) (\text{number of terms})
$$

$$
\left( 
    \dfrac{n + k}{2} 
\right) (n - k + 1)
$$

Thus:

$$
\sum_{k=1}^n \dfrac{(n + k)(n - k + 1)}{2}
$$

Factor out $\dfrac{1}{2}$:

$$
\dfrac{1}{2} \sum_{k=1}^n (n + k)(n - k + 1)
$$

---

## Simplify the Inner Expression

Expand the inner expression $(n + k)(n - k + 1)$:

$$
\begin{aligned}
    (n + k)(n - k + 1) &= n^2 - nk + n + kn - k^2 + k \\
                       &= n^2 + n - k^2 + k
\end{aligned}
$$

Substitute back into the sum:

$$
\dfrac{1}{2} \sum_{k=1}^n (n^2 + n - k^2 + k)
$$

Break into individual sums:

$$
\dfrac{1}{2} 
\left( 
    \sum_{k=1}^n n^2 + \sum_{k=1}^n n - \sum_{k=1}^n k^2 + \sum_{k=1}^n k 
\right)
$$

Simplify:

$$
\dfrac{1}{2} 
\left[ 
    n^3 + n^2 - \mathbf{S} + \dfrac{n(n+1)}{2} 
\right]
$$

Balance the equation:

$$
\mathbf{S} = \dfrac{1}{2} \left[
    n^3 + n^2 - \mathbf{S} + \dfrac{n(n+1)}{2} 
\right]
$$

$$
2\mathbf{S} = n^3 + n^2 - \mathbf{S} + \dfrac{n(n+1)}{2}
$$

$$
3\mathbf{S} = n^3 + n^2 + \dfrac{n(n+1)}{2}
$$

$$
3\mathbf{S} = \dfrac{2n^3 + 2n^2 + n^2 + n}{2}
$$

$$
\begin{aligned}
    \mathbf{S} &= \dfrac{2n(n^2 + n) + (n^2 + n)}{6} \\
    \mathbf{S} &= \dfrac{(n^2 + n)(2n + 1)}{6} \\
    \mathbf{S} &= \dfrac{n(n + 1)(2n + 1)}{6}
\end{aligned}
$$

Therefore:

$$
\boxed{ \sum_{k=1}^n k^2 = \dfrac{n(n + 1)(2n + 1)}{6} }
$$

## Additional Insights
From this we can see that also:

$$
\sum_{k=1}^n k^2 = \left( \dfrac{2n + 1}{3} \right) \left[ \dfrac{n(n + 1)}{2} \right]
$$

$$
\sum_{k=1}^n k^2 = \left( \dfrac{2n + 1}{3} \right) \sum_{k=1}^n k
$$


Therefore, the sum of squares is connected to the common arithmetic sum by a factor of $ \frac{2n + 1}{3} $

## Explore Other Methods (Comprehensive Understanding)

To deepen understanding, explore alternative methods:
- **Mathematical Induction**: Prove the formula by assuming it holds for $n$ and showing it holds for $n+1$.
- **Telescoping Series**: Use $(k+1)^3 - k^3 = 3k^2 + 3k + 1$ to derive the sum.

This comprehensive approach ensures a holistic grasp of the concept.
