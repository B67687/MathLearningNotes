## Derivation of Geometric Sum

For a geometric sum, it takes the form

$$
S_n = a + ar + ar^{2} + ar^{3} + \cdots + ar^{n-1}, \text{ for n terms}
$$

Where `a` is the constant factor, `r` is the constant multiple that acts on `a`

Which means it can also be presented in this form:

$$
S_n = a(1 + r + r^{2} + r^{3} + \cdots + r^{n-1})
$$

How do we derive the general form of the geometric sum?
- Well, notice that each term ratio to the next term is exactly `r`?
- Let us use this fact to set up a circumstance whereby we can cancel out almost all of the terms in between and just leave the first and last term

Let us focus on this part instead

$$
\textcolor{lightgray}{S_n = a(}1 + r + r^{2} + r^{3} + \cdots + r^{n-1}\textcolor{lightgray}{)}
$$

Let this be 

$$
Z = 1 + r + r^{2} + r^{3} + \cdots + r^{n-1}
$$

We can take itself, minus `r` of itself, thereby cancelling out basically every term in between, and introducing a new largest term

$$
\begin{align}
    Z \space (1-r) &= (1 + r + r^{2} + r^{3} + \cdots + r^{n-1}) \\
    &\qquad - (r + r^{2} + r^{3} + r^{4} + \cdots + r^{n})
\end{align}
$$

Here, we can cancel out the rest in between

$$
\begin{align}
    Z \space (1-r) &= (1 + \cancel{r} + \cancel{r^{2}} + \cancel{r^{3}} + \cdots + \cancel{r^{n-1}}) \\
    &\qquad - (\cancel{r} + \cancel{r^{2}} + \cancel{r^{3}} + \cancel{r^{4}} + \cdots + r^{n}) \\
    &= 1 - r^{n}\\
\end{align}
$$


$$
\boxed{
    Z = \dfrac{1 - r^{n}}{1-r}
}
$$

Notice that because of how the 2 series $\underline{1 + r + ... + r^{n-1}}$ and $\underline{r + r^{2} + ... + r^{n}}$ both share equal terms, we can also take the reverse difference between them


Such that 
$$
\begin{align}
    Z \space (r-1) &= (\cancel{r} + \cancel{r^{2}} + \cancel{r^{3}} + \cancel{r^{4}} + \cdots + r^{n}) \\
    &\qquad - (1 + \cancel{r} + \cancel{r^{2}} + \cancel{r^{3}} + \cdots + \cancel{r^{n-1}}) \\
    &= r^{n} - 1\\
\end{align}
$$


$$
\boxed{
    Z = \dfrac{r^{n} - 1}{r-1}
}
$$

Therefore

$$
\boxed{
    Z = \left( \dfrac{1 - r^{n}}{1-r} \right) \qquad \text{or} \qquad  Z = \left( \dfrac{r^{n} - 1}{r-1} \right)
}
$$

Therefore, substituting it back

$$
S_n = a(1 + r + r^{2} + r^{3} + \cdots + r^{n-1})
$$

$$
\boxed{
    S_n = a \left( \dfrac{1 - r^{n}}{1-r} \right) \qquad \text{or} \qquad S_n = a \left( \dfrac{r^{n} - 1}{r-1} \right)
}
$$

Or more normally seen as 

$$
\boxed{
    S_n = \dfrac{a(1 - r^{n})}{1-r} \qquad \text{or} \qquad S_n = \dfrac{a(r^{n} - 1)}{r-1}
}
$$
