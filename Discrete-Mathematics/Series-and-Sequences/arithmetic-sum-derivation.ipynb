# Understanding Arithmetic Series Sums Through Symmetry of Averages

The sum of an arithmetic series is a classic result in mathematics, often proven using the method attributed to Gauss, which pairs terms from the ends of the sequence. 
For example, to sum the numbers $1 + 2 + \cdots + 100$, Gauss paired terms from the ends ($1 + 100$, $2 + 99$, etc.) to derive the total

In this explanation, we present an alternative approach that reveals why this pairing works by focusing on the symmetry of averages.

This method emphasizes conceptual understanding and provides a fresh perspective on the arithmetic series sum, making it particularly suitable for educational settings where intuition is valued.

## The Key Idea: Averages as Symmetry on a Number Line

Let’s rethink what an average means for numbers in an arithmetic sequence, where each term increases by a common difference. Instead of just “sum divided by the number of terms,” we can see the average as the central point around which the numbers are symmetrically balanced.


Consider any term:
$$
\large\boxed{a}
$$

If we move a some distance `d` away from the term, we can get to 2 points
$$
\large{\boxed{a - d} \longleftarrow a \longrightarrow \boxed{a + d}}
$$

Notice that the way we built this means `a` is average of the 2 points, because we have moved the same amount both left and right with reference to `a`

If we keep moving the same amount both directions, `a` will keep being the average

Let's imagine the above as term itself `b`:

$$
\large{b : \boxed{
    a - d \longleftarrow a \longrightarrow a + d
    }
}
$$

Let's move `d` away from `b` from both sides

$$
\large{
    \boxed{b - d} \longleftarrow b \longrightarrow \boxed{b + d}
}
$$

... which is equivalent to this:

$$
\large{ \boxed{a - 2d} \longleftarrow \boxed{a - d \longleftarrow a \longrightarrow a + d} \longrightarrow \boxed{a + 2d}}
$$

$$
\large{ \boxed{a - 2d} \longleftarrow \boxed{a - d} \longleftarrow a \longrightarrow \boxed{a + d} \longrightarrow \boxed{a + 2d}}
$$

Notice because we can keep doing this, `a` is guaranteed to always be in the centre as we always move at **equal magnitudes** but in **opposite directions**

$$
\large{
    \dots \longleftarrow \boxed{a - d} \longleftarrow \boxed{a - d} \longleftarrow a \longrightarrow \boxed{a + d} \longrightarrow \boxed{a + 2d} \longrightarrow \dots
}
$$

It thus can also allow for pairs equidistant to the midpoint to not exist, because `d` can be smaller or bigger, the midpoint stays constant always regardless. 

$$
\large{
    \dots \longleftarrow \boxed{a - 2d} \textcolor{darkgray}{\longleftarrow \boxed{a - d}} \longleftarrow a \longrightarrow \textcolor{darkgray}{\boxed{a + d} \longrightarrow} \boxed{a + 2d} \longrightarrow \dots
}
$$

In essence it all depends on how small or big the chocice of `d` is.

Thus,

$$
\boxed{
    \text{The centre (and average), } \mathbf{a}, \text{ can always be derived from any 2 points equidistant from } \mathbf{a}
}
$$

From `a`:
$$
\dfrac{(a - 0) + (a + 0)}{2} = a
$$

$$
\dfrac{(a - d) + (a + d)}{2} = a
$$

$$
\dfrac{(a - 2d) + (a + 2d)}{2} = a
$$

$$
\vdots
$$

$$
\dfrac{\boxed{a - (n - 1)d} + \boxed{a + (n - 1)d}}{2} = a
$$

But equally, any 2 points equidistant from the furthest points

$$
\dfrac{\boxed{a - (n - 1)d} + \boxed{a + (n - 1)d}}{2} = a
$$

$$
\vdots
$$

$$
\dfrac{(a - d) + (a + d)}{2} = a
$$

$$
\dfrac{(a - 0) + (a + 0)}{2} = a
$$

Since the first and last points always exist, we can always find the average of the whole sum by taking the average of the first and last points
- Even for sums with 1 term, whereby the first and last points are itself


Therefore:

$$
\begin{align}
    \text{Average of this series } &= \text{Average of the first and last term } \\
    &= \dfrac{\text{first } + \text{ last}}{2}
\end{align}
$$

The total of this series is:

$$
\text{Total } = \text{Average } \cdot \text{ Number of Terms}
$$

Therefore

$$
\boxed{\text{Total } = \dfrac{\text{first } + \text{ last}}{2} \cdot \text{ Number of Terms}}
$$

# Detailed Explanation of the Arithmetic Series Sum Using Symmetry of Averages

Now knowing this concept, we can apply it to the arithmetic sum


## Rebuilding any Arithmetic Sequence
Consider the simple case:
$$
\boxed{a}
$$

In an arithmetic sequence, there is a common difference of `d`, 

Normally the sequence is built in one direction based off the base term `a`.

In sum `S` with `n` terms:

$$
\begin{align}
    &S_1 = a \\
    &S_2 = a + \boxed{a + d} \\
    &S_3 = a + \boxed{a + d} + \boxed{a + 2d} \\
    &S_n = a + \boxed{a + d} + \boxed{a + 2d} + \dots + \boxed{a + (n - 1)d}
\end{align}
$$

An equal way of viewing the arithmetic sequence is to build from both sides:

$$
\begin{array}{cc}
    S_1 = & a \\
    S_3 = & \boxed{a - d} + a + \boxed{a + d} \\
    S_5 = & \boxed{a - 2d} + \boxed{a - d} + a + \boxed{a + d} + \boxed{a + 2d} \\
    S_n = & \boxed{a - \left( \dfrac{n - 1}{2} \right) d} + \dots + \boxed{a - 2d} + \boxed{a - d} + a + \boxed{a + d} + \boxed{a + 2d} + \dots + \boxed{a + \left( \dfrac{n - 1}{2} \right) d} \\
\end{array}
$$


It seems that I am missing the series where there are even terms

While conserving this structure, we can get the series of even terms:

$$
\begin{array}{cc}
    S_2 = & \textcolor{lightgray}{a} \\
    S_2 = & \boxed{a - d} + \textcolor{lightgray}{a} + \boxed{a + d} \\
    S_4 = & \boxed{a - 2d} + \boxed{a - d} + \textcolor{lightgray}{a} + \boxed{a + d} + \boxed{a + 2d} \\
    S_n = & \boxed{a - \left( \dfrac{n}{2} \right) d} + \dots + \boxed{a - 2d} + \boxed{a - d} + \textcolor{lightgray}{a} + \boxed{a + d} + \boxed{a + 2d} + \dots + \boxed{a + \left( \dfrac{n}{2} \right) d} \\
\end{array}
$$

Notice that the average doesn't exist as a term in sums with even number of terms

Therefore, for any number of terms, the average will always be `a`, and therefore the total will be easily calculated as:

$$
\text{Arithmetic Sum } = \text{Average } \cdot \text{Number of Terms}
$$

Or like this:

$$
\text{Arithmetic Sum } = a \cdot n
$$

And since the average can always be calculated from the `first` and `last` terms

$$
\boxed{\text{Arithmetic Sum } = \dfrac{\text{first } + \text{ last}}{2} \cdot \text{ Number of Terms}} 
$$


**Thus**:

Let an arithmetic sequence be defined with:
- **First term**: $a$
- **Common difference**: $d$
- **Number of terms**: $n$
- **Terms**: $a, a + d, \ldots, a + (n-1)d$
- **Sum**: $S_n = a + (a + d) + \cdots + (a + (n-1)d)$

$$
\begin{align}
    S_n &= a \cdot n \\ 
    &= \dfrac{\text{first} + \text{last}}{2} \cdot n \\
    &= \dfrac{a + [ a + (n-1)d ]}{2} \cdot n \\
    &= \dfrac{2a + (n-1)d}{2} \cdot n \\
    &= \dfrac{n}{2} [ 2a + (n-1)d ]
\end{align}
$$

Therefore, we get to our familiar formula for the arithmetic sum:
$$
\boxed{S_n = \dfrac{n}{2} [ 2a + (n-1)d ]}
$$


## Application to the Sum of Arithmetic Sequence with Common Difference 1

$\Huge{
S_1 = 1 + 2 + 3 + \dots + n
}$

### Applying the formula
Now let's find its formula using the knowledge of Symmetric Averages

The sequence $1, 2, 3, \dots, n$ has:
- First term: $1$
- Last term: $n$
- Number of terms: $n$

The sum of an arithmetic sequence is given by:

$$
\text{Total} = \text{Average} \cdot \text{Number of Terms}
$$

Calculate the average of the first and last terms:

$$
\text{Average} = \dfrac{\text{First} + \text{Last}}{2} = \dfrac{1 + n}{2}
$$

The number of terms is:

$$
\text{Number of terms} = n
$$

So the total sum is:

$$
\text{Total} = \left( \dfrac{1 + n}{2} \right) \cdot n = \dfrac{n(n + 1)}{2}
$$

Therefore:

$$
\boxed{S_1 = \dfrac{n(n + 1)}{2}}
$$

# Gauss' Pair-Matching
#### And thus this is why Gauss' Pair-Matching method works

---
Write the sequence forward:

$$
S_n = 1 + 2 + 3 + \dots + (n - 1) + n
$$
Write the sequence in reverse:

$$
S_n = n + (n - 1) + \dots + 3 + 2 + 1
$$

Adding the two sequences term by term pairs the terms equidistant from the average:

$$
2S_n = (1 + n) + (2 + (n - 1)) + \dots + ((n - 1) + 2) + (n + 1)
$$

Each pair (as observed as well as proven) sums to the same value: 

$$
\textcolor{lightgray}{2S_n =} (1 + n) + (n + 1) + \dots + (n + 1) + (n + 1)
$$

There are $n$ terms in the original sequence, so there are $n$ pairs:

$$
2S_n = \underbrace{(n + 1) + (n + 1) + \dots + (n + 1) + (n + 1)}_{\text{n pairs}}
$$

$$
2S_n = n \cdot (n + 1)
$$

Divide by 2:

$$
S_n = \dfrac{n \cdot (n + 1)}{2}
$$

Thus:
$$
\boxed{S_n = \dfrac{n(n + 1)}{2}}
$$
